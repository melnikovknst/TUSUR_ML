#Шаг №3:
Создание датафрейма на основе файла products_table.csv

```python
df = pd.read_csv('/content/drive/MyDrive/ML/TUSUR_ML/pandas_basics/data/products_table.csv', index_col=0)
df
```

#Шаг №4:
Создать столбец "Цена со скидкой", вычислив цену товара с учетом скидки.

```python
df['Цена со скидкой'] = df['Цена'] * (1 - df['Скидка в процентах'] / 100)
df
```

#Шаг №4:
Отсортировать товары по убыванию цены со скидкой и вывести топ-10 самых дорогих товаров.

```python
sorted_df = df.sort_values('Цена со скидкой', ascending = False)
sorted_df.head(10)
```

#Шаг №5:
Сгруппировать данные по категориям и вычислить среднюю цену товара (как с учетом скидки, так и без неё) по каждой категории.

```python
grouped_df = df.groupby('Категория')
agg_df = grouped_df.agg({'Цена' : 'mean', 'Цена со скидкой' : 'mean'})
agg_df
```
