# Описание подготовки данных

Подготовка данных — это процесс приведения данных в формат, подходящий для анализа или использования в моделях машинного обучения. Этот процесс включает несколько шагов, направленных на то, чтобы данные были корректными, полными и качественными, и их можно было использовать с моделями машинного обучения.

Например, если вы возьмете базовый алгоритм классического машинного обучения и попытаетесь передать ему сырые данные, то в лучшем случае увидите достаточно плохие метрики. В худшем - алгоритм сразу выкинет ошибку из-за того, что на вход ему подаете слова, а не числа. А алгоритмы работают именно с числами.

В подготовке с данными выделяют несколько шагов, которые максимально стараются захватывать разные аспекты работы с данными. Однако для разных типов данных могут быть модификации.

0. Сбор данных
1. Первичная очистка
2. EDA
3. Очистка данных
4. Feature Engineering

Нулевой шаг, или этап перед подготовкой данных - это всегда сбор и агрегация данных, так как нужен какой-то банк данных, с которым будем работатью

В первый шаг можно выделить первичную очистку. При агрегации данных может возникнуть много разных исключительных ситуаций, и хоть дата инженеры стараются собрать данные качественно, проверить данные все же стоит - например, соотнести имеющиеся типы данных с требуемыми (например, чтобы в числовой колонке не было строк). Также данные можно проверить на опечатки, регистр в словах, явные полные дубликаты строк.

Второй - это разведочный анализ. О нем подробнее поговорим чуть позже, но главное, что он нам дает - это статистическое описание данных и первые выводы по ним.

В непосредственно очистке данных мы занимаемся чисткой артефактов, которые нашли в процессе EDA и теперь можем исправить, - это работа с аномалиями, пропусками, удаление вновь возникших дубликатов.

И заключительный этап - это feature engineering. На самом деле этот этап содержит в себе другие подъэтапы, о которых тоже поговорим дальше, но здесь мы создаем новые признаки, кодируем, преобразовываем данные, можем их расширить или сократить.

# Сбор данных

**Датасет (dataset)** — это набор данных, который собирается и систематизируется для анализа, обучения моделей машинного обучения, исследований или других целей. Датасет может содержать различные типы данных, такие как числовые, текстовые, категориальные, временные ряды и т.д.

В рамках машинного обучения используют данные разнообразных типов, отталкиваясь от того, какие данные и в каком формате есть у бизнеса, какие данные принимают различные модели и архитектуры и так далее. Их можно разделить на следующие категории.

* Таблицы наблюдений.	Каждый объект в таблицах наблюдений (матрицах объекты – признаки) описывается набором своих характеристик, называемых признаками. Признаки могут быть числовыми или нечисловыми. Каждому объекту соответствует либо метка класса, либо некоторое число. Хранятся такие данные чаще всего в формате электронных таблиц – csv, xslx и т.п.

* Временные ряды или сигналы.	Временные ряды или сигналы – это последовательность некоторых измерений во времени. Измерение представляет собой число или вектор и может быть преобразовано в признаковое описание объекта в заданный отсчет или интервал времени.
Форматы после извлечения признаков – csv, xslx и т. д. Исходные форматы могут быть самыми разнообразными.

* Изображения и видео.	Машинное обучение широко применяется для распознавания образов и идентификации объектов на изображениях и видеороликах для восстановления, изменения и генерации визуального материала на основе существующих данных, а также для множества других подобных задач. Для достижения высокого качества алгоритмы обучаются на большом количестве изображений и видео огромного объема.
Форматы – растровая и векторная графика (.png, .jpeg, .gif, tiff, .bmp, .cdr, .eps, .svg, .wmf и т. д.), видео (.mp4, .mov, .wmv, .avi. и др.)

* Тексты.	Текстовые данные необходимы для обучения моделей, целями построения которых являются: распознавание смысла текста, генерация ответов на вопросы пользователей, оценка эмоциональной окраски сообщений (например, в социальных сетях или интернет-магазинах), поиск необходимой информации в тексте, фильтрация спама и другие задачи обработки естественного языка

* Другие типы.	Данные могут быть представлены в виде графов, кода, результатов запросов к базе данных, и т. д. Как правило, такие данные приводятся к табличному типу путем предварительной обработки данных и извлечения признаков. Нередко в рамках одной и той же задачи требуется оперировать данными различных типов. Такие наборы данных называют смешанными

## Таблицы наблюдений

Таблицы наблюдений считаются стандартным способом представления исходных данных в задачах обучения с учителем – классификации и регрессии. Большинство алгоритмов машинного обучения приспособлено к обработке именно таких данных.

В таблицах наблюдений каждая строка описывает некоторый объект. Объект называют экземпляром или образцом данных (в англоязычной литературе чаще применяют термин instance). Столбцы – это признаки (атрибуты) объекта.

Признак (feature) – результат измерения некоторой характеристики изучаемого объекта.

Помимо признакового описания у каждого объекта имеется выходная переменная, если рассматриваем задачу обучения с учителем. Обычно это один столбец, в котором находится либо число из некоторого непрерывного диапазона (задача регрессии), либо метка класса из некоторого конечного множества классов (задача классификации).

**ПРИМЕР**

Если имеется задача диагностики какого-либо заболевания, то каждая строка таблицы наблюдения – это пациент, а признаками будут являться различная информация о пациенте – анамнез, результаты анализов и другие важные для определения заболевания данные. Или, например, как на картинке — классический датасет ирисов Фишера. Первые четыре колонки - замеры лепестков и соцветия ирисов, а последняя — выходная переменная, вид ириса.



## Типы признаков
В машинном обучении данные можно разделить на две основные категории: количественные (или числовые) и категориальные данные.

Числовые (количественные) данные — это как когда у нас есть линейка, весы или любая другая шкала, чтобы измерить что-то. Эти данные можно выразить в числах. Они делятся на две подгруппы:

* Непрерывные данные (Continuous Data): Это данные, которые могут принимать любое значение в пределах определенного диапазона. Примеры включают рост, вес, температуру, время и т.д. Эти данные могут быть дробными и обладают бесконечным числом возможных значений в заданном интервале.

* Дискретные данные (Discrete Data): Это данные, которые могут принимать только определенные значения. Обычно это целые числа. Примеры включают количество детей в семье, количество дефектов на продукте и т.д.

Категориальные данные представляют собой характеристики или качественные атрибуты, которые не имеют числового значения.
Рассмотрим подвиды категориальных данных:
* бинарные (bool) признаки удобно использовать, чтобы указать на наличие или отсутствие какой-либо характеристики у объекта. Чаще всего в таблицах бинарные значения представляют цифрами 0 и 1, реже встречаются текстовые пары True-False;
* номинальные признаки задаются некоторым конечным множеством значений. Обычно это строковые переменные или числовые коды;
* порядковые признаки являются конечным упорядоченным множеством, где между элементами имеется иерархия;


Отдельно вне этих двух групп выделают иные данные: тексты, списки, коды и другие редкие форматы.

image.png

Понимание разницы между количественными и категориальными данными важно, потому что это помогает нам правильно их использовать в машинном обучении.

* Количественные данные позволяют моделям находить числовые зависимости. Например, модель может понять, что рост и вес человека связаны между собой.
* Категориальные данные помогают моделям различать категории. Например, модель может понять, что фрукты могут быть разного типа и каждый тип по-своему важен.

Представьте, что мы строим модель для предсказания успеваемости студентов. У нас есть количественные данные, такие как количество часов, потраченных на учёбу, и категориальные данные, такие как уровень образования родителей (начальное, среднее, высшее). Используя оба типа данных, мы можем сделать более точные и полезные предсказания.

Также на этапе обработки данных важно понимать типы признаков, поскольку от этого зависит способ обработки пропущенных значений, кодирования и других операций. Например, если в ваших данных дискретный признак «количество детей в семье» имеет пропуски в некоторых записях, то, вероятно, для вменения пропусков лучше использовать моду, чем среднее арифметическое? Среднее значение редко бывает целочисленным, и ведь не может на семью приходиться 1.73 ребенка.

## Источники данных

В современном мире данные – это ресурс, представляющий собой немалую ценность. Для многих компаний, имеющих собственные команды для анализа данных и машинного обучения, не так страшно раскрыть подробности своих моделей и алгоритмов, как расстаться с данными.

Тем не менее, иногда коммерческие компании, государственные структуры и университеты публикуют данные в открытом доступе. Причиной тому скорее всего служат попытки найти новые идеи, чем следование тренду на открытость информации. Искать такие данные стоит либо на интернет-ресурсах непосредственных владельцев данных, либо в репозиториях.

Существуют категории общедоступных данных, сбором и обработкой которых может заниматься любой желающий. Это данные, публикуемые в СМИ, а также выкладываемые пользователями данные из социальных сетей.

Источниками данных также являются научные публикации. В крупных журналах существуют требования публиковать все данные, накопленные в рамках исследования, однако далеко не все такие журналы являются бесплатными. Впрочем, существуют научные коллективы, поддерживающие идею открытости информации и раскрывающие свои данные бесплатно, даже если журналы этого не требуют. Сырые данные из каких-либо определенных сфер можно найти в соответствующих тематических журналах. В изданиях, посвященных анализу данных и машинному обучению, можно найти даже готовые к использованию наборы.

Репозиторий – место для хранения каких-либо данных.

В рамках данного модуля нас интересуют репозитории, позволяющие обмениваться наборами данных, пригодными для анализа и построения моделей машинного обучения.

Рассмотрим наиболее известные среди них.

**UCI Machine Learning Repository**

Репозиторий UCI является одним из самых ранних репозиториев с открытым доступом (создан в 1987 г.). Оригинальная версия ресурса не обновляется с 2021 г., но существует beta-версия UCI, отличающаяся более современным интерфейсом.

Скачивание файлов с репозитория доступно без регистрации. Наиболее популярные наборы данных сопровождаются большим количеством полезного материала для исследователей – описанием содержимого, ссылками на статьи и даже результатами построения базовых моделей машинного обучения.

**Kaggle**

Самой распространенной платформой для специалистов по машинному обучению на сегодняшний момент является Kaggle.

Популярность этой платформы объясняется тем, что кроме загрузки и скачивания наборов данных пользователи могут найти на ней много полезного: соревнования по построению моделей машинного обучения, полезные статьи и уроки, обсуждения опубликованного кода и т. д. Ресурс имеет встроенную среду программирования на языке Python, так что модели к опубликованным данным можно строить прямо на Kaggle. У размещенных в открытом доступе наборов данных, решений и пользователей есть оценки и рейтинги, благодаря чему новым пользователем проще ориентироваться в контенте.

Таким образом, Kaggle – отличный ресурс для практики своих умений в машинном обучении. Для использования платформы требуется регистрация.

## **Сбор данных из открытых источников**
Собирать данные можно и вручную, но большой массив информации таким способом собрать трудно. Разумнее обратиться к автоматизированным методам – краулингу, скрапингу и парсингу.

* Краулинг – исследование веб-страниц для индексации контента по заданным URL-адресам, считывание и накопление вложенных ссылок, а также сведений о содержимом этих страниц.

Краулингом в первую очередь занимаются поисковые боты компаний, которые выполняют поиск и выдачу данных, – Google, Yandex, Yahoo, Bing и т. д. Поисковые боты (скрипты, программы) обрабатывают веб-ссылки из списка заданных URL-адресов. Краулер просматривает все страницы веб-сайта, переходит по ссылкам, находит новые страницы и извлекает контент относительно неизбирательным образом. Процесс работы краулера можно сравнить с составлением каталога библиотеки.

* Скрапинг – обход веб-страниц по заданным URL-адресам и извлечение данных.

* Парсинг – извлечение необходимой информации из неструктурированных данных и ее преобразование к нужному виду и формату.

Скрапинг и парсинг в русском языке разделяют довольно редко. Чаще термином «парсинг» обозначают оба этих понятия.

Скрапинг подразумевает считывание данных с определенного источника (преимущественно с интернет-ресурсов) и их сохранение в виде, удобном для автоматизированной обработки. В процессе скрапинга данные очищаются от лишнего кода с целью получения только нужных текстов, изображений или видео.

Парсинг направлен на извлечение из имеющегося массива данных конкретной информации, из которой формируется файл заданного вида.

**ПРИМЕР**

Допустим, вы решили работать аналитиком данных на фрилансе и вам необходимо решить, какую стоимость установить для своих услуг.

С помощью поискового бота-краулера вы составляете перечень всех сайтов, где предоставляются услуги фриланса.

Далее вы программируете скрапинг-скрипт для найденных сайтов, который экспортирует в формат txt все объявления о предоставлении или запросу услуг по анализу данных за последние пару лет.

Напоследок вы пишите парсер, чтобы из txt сформировать электронную таблицу с признаками «стоимость», «дата», «вид работ» и «популярность объявления».

После этого вы загружаете свое резюме на все найденные вами ресурсы, указав в резюме умение виртуозно собирать данные.

Проводить простейший скрапинг можно через браузер. Кроме того, на настоящий момент существует немало готовых программ, которые предлагают возможности и краулинга, и скрапинга, и парсинга, и всевозможного экспорта данных, их визуализации и даже анализа. Но нужно понимать, что они предлагают универсальные решения. Для проведения нестандартного и избирательного сбора данных стоит специально разрабатывать соответствующий код.

## **Юридическая сторона сбора данных**
Как уже упоминалось ранее, данные – это ресурс. Даже при использовании данных из репозиториев нужно помнить, что они кому-то принадлежат. А при самостоятельном сборе данных, разумеется, следует обращать особое внимание на авторские права.

Каждый авторитетный репозиторий на странице набора данных приводит информацию об авторах данных и о лицензии, которая регламентирует доступ к данным.

Главное, что стоит помнить – при скачивании данных с репозиториев необходимо проверять и раздел с лицензией (чаще всего в нем можно ознакомиться с подробным описанием всех условий использования данных), и описание данных. Нередко в описании авторы указывают, на какой именно ресурс или статью требуется сделать ссылку при использовании данных.

Что касается парсинга интернет-ресурсов, в этом процессе нет ничего подозрительного или незаконного. Однако это не означает, что любой вид парсинга и скрапинга является законным. Как и всякая человеческая деятельность, они должны оставаться в определенных границах. Наиболее важными границами здесь являются личные данные и права интеллектуальной собственности.

Резюмируя вышесказанное, следует отметить следующее:

- законным является парсинг данных, для доступа к которым с целью проведения научных исследований или самообучения не нужно проходить аутентификацию;
- при исследовательской работе с персональными данными необходимо следовать документам, посвященным охране персональных данных;
- при парсинге данных для коммерческого использования (с целью извлечения прибыли) требуется ознакомиться и с требованиями законодательства, и с конкретными условиями использования информации с ресурсов, которые являются источниками данных.

## Немного вспомним тервер и матстат

**Распределение** в статистике и теории вероятностей — это функция, которая описывает, как значения случайной переменной распределяются или "расползаются" по возможным результатам. Оно показывает, с какой вероятностью случайная переменная принимает определённые значения или как часто эти значения встречаются в наборе данных.

**Функция плотности вероятности** показывает, какова вероятность того, что случайная величина примет определённое значение в бесконечно малом интервале. График функции плотности часто называют кривой распределения. Представьте график, на котором по оси X откладываются возможные значения, а по оси Y — их «плотность» или вероятность того, что случайно полученное значение будет рядом с этой точкой. Высота графика в каждой точке указывает на "относительную вероятность" наблюдения значения из окрестности этой точки. Чем выше график в определённой области, тем больше вероятность того, что случайная величина примет значение в этой области. Иначе говоря, пики — области с наиболее вероятными значениями. Это полезно, например, при анализе данных, чтобы понять, где "сконцентрированы" основные вероятности появления различных значений. Для нормального распределения эта кривая имеет форму колокола.

**Функция распределения** показывает вероятность того, что случайная величина примет значение, меньшее или равное данному. Она растёт от 0 до 1 по мере увеличения значений случайной величины.

Распределение помогает понять, как ведут себя данные. Зная распределение, можно делать предположения о вероятностях событий, рассчитывать средние значения, разбросы и другие параметры.



Также выделяют **скошенные распределения** (или асимметричные распределения) — это тип распределений, в которых данные неравномерно распределены вокруг среднего значения. В отличие от симметричных распределений (например, нормального распределения), в которых левая и правая части графика одинаковы, в скошенных распределениях одна "сторона" (хвост) может быть длиннее или более выраженной, а большинство значений сосредоточено около пика.

Скошенные распределения часто связаны с наличием выбросов — экстремальных значений, которые могут искажать выводы моделей. Например, в скошенном распределении высокие выбросы могут иметь большое влияние на среднее значение, делая его менее репрезентативным для большинства данных.

Если данные не нормальны, можно попробовать их преобразовать, чтобы сделать их распределение ближе к нормальному. Для этого используют лог-преобразование, преобразование Бокса-Кокса и другие.

**Модальность** — это количество "пиков", мод в распределении:

* Унимодальное распределение имеет один пик (моду), как у нормального распределения.
* Бимодальное распределение имеет два пика, что может указывать на наличие двух различных групп в данных. Пример: распределение роста, если в выборке есть как взрослые мужчины, так и женщины — каждый пол может образовывать свой пик.
* Мультимодальное распределение имеет более двух пиков, что может говорить о наличии нескольких подгрупп в данных.

**Меры центральной тенденции**

Обычно мы хотим иметь некое представление о том, где наши данные центрированы. Чаще всего для этих целей используется **среднее** (или среднее арифметическое) значение, которое берется как сумма данных, деленная на их количество.

Для двух точек средней является точка, лежащая посередине между ними. По мере добавления других точек среднее значение будет смещаться в разные стороны в зависимости от значения каждой новой точки. Например, если у вас 10 точек данных, и вы увеличиваете значение любой из них на 1, то вы увеличиваете среднее на 0.1.

Мы также иногда будем заинтересованы в **медиане**, которая является ближайшим к центру значением (если число точек данных нечетное) либо средним арифметическим, взятым как полусумма двух ближайших к середине значений (если число точек четное). Иначе говоря, медиана — это такое значение в упорядоченном наборе данных, которое делит его на две равные части: 50% значений меньше медианы, и 50% значений больше медианы. Если у вас есть нечетное количество друзей и вы выстроите их в ряд по росту, медиана - это рост друга, стоящего в середине.

Обратите внимание, что медиана - в отличие от среднего - не зависит от каждого значения в наборе данных. Например, если сделать наибольшую точку еще больше (или наименьшую точку еще меньше), то срединные точки останутся неизменными, следовательно, и медиана тоже не изменится.

Также для дискретных величин выделяют **моду** — это значение, которое встречается чаще всего. Если у вас на вечеринке большинство людей предпочитают одну и ту же песню, эта песня - мода.

**Меры разброса данных**

Вариация служит мерой разброса наших данных. Как правило, это статистические показатели, у которых значения, близкие к нулю, означают полное отсутствие разброса, а большие значения (что бы это ни означало)- очень большой разброс.

Например, самым простым показателем является **размах**, который определяется как разница между максимальным и минимальным значениями данных. Размах равен нулю, когда max и min эквивалентны, что происходит только тогда, когда все элементы х равны между собой, и значит, разбросанность в данных отсутствует. И наоборот, когда размах широкий, то max намного больше min, и разбросанность в данных высокая.
Как и медиана, размах не особо зависит от всего набора данных. Набор данных, все точки которого равны 0 или 100, имеет тот же размах, что и набор данных, чьи значения представлены числами 0, 100 и большим количеством чисел 50 между ними, хотя кажется, что первый набор "должен" быть разбросан больше.

Более точным показателем вариации является **дисперсия** — это средний квадрат отклонений значений от их среднего. Она показывает, насколько значения разбросаны вокруг среднего. Чем больше дисперсия, тем больше разброс.

При этом в каких бы единицах ни измерялись данные (в "друзьях", например), все меры центральной тенденции вычисляются в тех же самых единицах измерения. Аналогичная ситуация и с размахом. Дисперсия же измеряется в единицах, которые представляют собой квадрат исходных единиц ("друзья в квадрате"). Поскольку такие единицы измерения трудно интерпретировать, то вместо дисперсии мы будем чаще обращаться к **стандартному отклонению** (корню из дисперсии): оно показывает, насколько сильно значения отличаются от среднего в среднем. Малое стандартное отклонение указывает на то, что значения близки к среднему, тогда как большое стандартное отклонение показывает, что значения распределены более широко.

**Что такое квартили и процентили?**

**Процентиль** — это статистическая мера, которая определяет значение в упорядоченном наборе данных, ниже которого находится определённый процент данных. Например, $p$-ый процентиль — это такое значение, ниже которого находится $p$% наблюдений в выборке. Каждый процентиль показывает, какой процент значений в выборке меньше или равен определённому значению.

Например:

* 10-й процентиль — это значение, ниже которого находится 10% всех данных.
* 90-й процентиль — это значение, ниже которого находится 90% данных.

* Медиана — это 50-й процентиль, точка посередине набора данных. Половина значений находится ниже медианы, а половина выше. Если у вас есть 100 человек, то рост 50-го человека в упорядоченном списке будет медианой.

  1. Первый квартиль (Q1) — это 25-й процентиль. 25% данных меньше этого значения.
  2. Медиана (Q2) – 50-й процентиль.
  3. Третий квартиль (Q3) — это 75-й процентиль. 75% данных меньше этого значения.

## Про нормальное распределение и почему это важно

**Нормальное распределение** (также известное как гауссово распределение) — это один из самых фундаментальных и широко используемых типов распределений вероятностей в статистике и машинном обучении. Его график напоминает симметричную "колоколообразную" кривую, где среднее, медиана и мода совпадают.

Основные характеристики нормального распределения:

1. Симметричность: Оно симметрично относительно среднего значения $\mu$, и при это мода, медиана и среднее совпадают.
2. Распределение значений, или правило трёх сигм:
* 68% данных находятся в пределах одного стандартного отклонения от среднего $(μ−σ,μ+σ)$
* 95% данных находятся в пределах двух стандартных отклонений $(μ−2σ,μ+2σ)$
* 99.7% данных находятся в пределах трёх стандартных отклонений $(μ−3σ,μ+3σ)$
3. Колоколообразная кривая: Пик на графике находится в точке среднего значения, а "хвосты" сужаются по мере удаления от него. Смысл нормального распределения становится понятен из его формы. Наиболее вероятные значения расположены вблизи его пика. По мере удаления от него, вероятность значений уменьшается и если значение расположено в «хвосте» распределения, то оно очень маловероятно.

**Почему нормальное распределение важно в машинном обучении?**
> Центральная предельная теорема утверждает, что сумма или среднее большого количества независимых случайных переменных с произвольными распределениями будет приближаться к нормальному распределению. Это значит, что многие явления в природе и в данных могут быть приближены нормальным распределением, если наблюдений достаточно много. Иначе говоря, это объясняет, почему нормальное распределение так часто встречается в природе и социальных явлениях — многие сложные процессы являются результатом суммирования множества независимых факторов. Поэтому модели и алгоритмы часто предполагают, что данные или ошибки имеют нормальное распределение.

Многие алгоритмы машинного обучения предполагают или работают лучше, если данные распределены нормально.

## Анализ от одной переменной (Univariate analysis)

Univariate анализ (анализ одной переменной) — это базовая форма статистического анализа данных, при которой рассматривается только одна переменная. Основная цель этого анализа — понять свойства и распределение данных в пределах одной переменной без учета взаимодействия с другими переменными.

В контексте univariate анализа, исследователи обычно интересуются такими параметрами, как:
1. Оценка распределения переменной — анализ помогает понять, как значения переменной распределены. Это может быть нормальное распределение, равномерное, экспоненциальное или любое другое, скошенное или нет, наличие длинных хвостов.

2. Вычисление статистических показателей: мода, медиана, среднее, дисперсия, стандартное отклонение.

3. Составление различных графиков: гистограммы, круговые диаграммы, графики распределения, бокс-плоты и другие.

Univariate анализ используется как первый шаг в исследовании данных перед тем, как перейти к более сложным методам, таким как корреляционный анализ или моделирование, и дает первое представление о нюансах в данных — наблюдаются ли какие-то особенности, например, возраста целевой аудитории, наиболее продаваемые категории товаров, даты пиков продаж и так далее.

## Анализ от двух и более переменных (Bivariate/Multivariate Analysis)

Bivariate анализ (анализ двух переменных) — это шаг дальше по сравнению с univariate анализом. В bivariate анализе исследуется взаимоотношение двух переменных для выявления их взаимосвязи или зависимости. Это помогает понять, как одна переменная изменяется относительно другой. Цель такого анализа — найти взаимосвязь между переменными, которая может быть линейной или нелинейной, сильной или слабой, положительной или отрицательной. Так можно находить зависимости не только между двумя переменными, но и разные инсайты, например, особенно сильное влияние на одну из категорий переменной или на определенный диапазон.

Задачи на этом этапе:

1. Исследование взаимосвязи между переменными: определить, существует ли связь между двумя переменными, и если да, то какова её природа (линейная, нелинейная).

2. Оценка направления и силы зависимости: выяснить, как изменяется одна переменная при изменении другой (например, рост продаж в зависимости от увеличения рекламы).

3. Оценка того, насколько сильна эта связь (сильная, средняя, слабая).

Способы анализа:
1. Вычисление различных коэффициентов корреляции
2. Построение графиков для двух переменных
3. Регрессионный анализ
4. Статистические тесты и другие методы




## Поиск закономерностей в данных

### Корреляционный анализ

Взаимосвязь двух переменных проявляется в совместной вариации: при изменении одного показателя имеет место тенденция изменения другого. Такая взаимосвязь называется корреляцией, а раздел статистики, который занимается взаимосвязями – корреляционный анализ.

Линейность корреляции проявляется в том, что точки расположены вдоль прямой линии. Положительный или отрицательный наклон такой линии определяется направлением взаимосвязи. Положительная корреляция в таких условиях — это такая связь, при которой увеличение одной переменной связано с увеличением другой переменной, а отрицательная — наоброт, увеличение одной переменной связано с уменьшением другой. Чем теснее (сильнее) взаимосвязь, тем ближе к прямой точки на диаграмме. Соответственно, чем слабее, тем сильнее данные рассеяны.

На картинке примеры связей: оси Х и Y, обозначающие признаки, точки - значения этих признаков для каждой записи.

Корреляция говорит нам только о том, насколько сильно пара количественных переменных линейно связана. Она не объясняет, как и почему. В то же время, отсутствие корреляции между двумя величинами ещё не значит, что между ними нет никакой связи. Например, зависимость может иметь сложный нелинейный характер, который корреляция не выявляет.

## Линейная корреляция

Корреляция Пирсона, также называемая коэффициентом корреляции, используется для измерения силы и направления (положительного или отрицательного) линейной связи между двумя количественными переменными.
Показатели линейной корреляции обычно представляют собой значение от -1 до 1, где:
* 0 означает отсутствие связи,
* ближе к 1 - положительная (за увеличением одной переменной увеличивается другая),
* к -1 - отрицательная (с увеличением одной другая уменьшается).



Простой (между двумя случайными величинами) коэффициент линейной корреляции (Пирсона) вычисляется по формуле:
\begin{align} r = \frac{\sum_{i=1}^n (x_n - \overline{x})(y_n-\overline{y})}{n\sigma_x \sigma_y}, \end{align}

где $n -$ число наблюдений (записей),

$x, y$ — случайные величины (например, два признака или признак и целевая переменная),

$\sigma_x, \sigma_y$ - стандартное отклонение случайных величин $x, y$.


Интерпретация коэффициента корреляции:

$0.9 \le |r| \le 1$ — очень сильная корреляция

$0.7 \le |r| < 0.9$ — сильная корреляция

$0.5 \le |r| < 0.7$ — умеренная корреляция

$0.3 \le |r| < 0.5$ — слабая корреляция

$0.0 \le |r| < 0.3$ — очень слабая корреляция

Очевидно, что если корреляция между переменными высокая, то, зная поведение входной переменной, проще предсказать поведение выходной, и полученное предсказание будет точнее (говорят, что входная переменная хорошо «объясняет» выходную). Однако чем выше корреляция наблюдается между переменными, тем очевиднее связь между ними, например, взаимозависимость между ростом и весом людей. Явление сильной линейной корреляции или полной линейной зависимости между признаками называют **мультиколлинеарностью**.

Чем плоха мультиколлинеарность? Во-первых, высокая корреляция между независимыми переменными делает невозможным определить, какая именно из них вносит вклад в предсказание зависимой переменной. Во-вторых, мультиколлинеарность негативно влияет на модели машинного обучения. Так, например, для линейных моделей она может приводить к неустойчивости коэффициентов.


Про другие подходы к расчету корреляции и отбору признаков поговорим на практике.

### Парадокс Симпсона

**Парадокс Симпсона** — это интересное и важное явление в статистике, которое демонстрирует, как тенденции, наблюдаемые в нескольких группах данных, могут меняться или даже исчезать при объединении этих групп.

**Простой пример**

Представьте, что мы хотим выяснить, какой из двух лекарств, A или B, эффективнее лечит болезнь. Мы проводим два отдельных эксперимента:

В первом эксперименте участвуют молодые пациенты.
Во втором эксперименте участвуют пожилые пациенты.
Результаты могут быть такими:

1. В группе молодых пациентов лекарство A помогает в 80% случаев, а лекарство B — в 70%.
2. В группе пожилых пациентов лекарство A помогает в 60% случаев, а лекарство B — в 50%.

На первый взгляд кажется, что лекарство A эффективнее в обеих группах. Но теперь объединим все данные вместе.

Допустим, в группе молодых пациентов было 10 человек, а в группе пожилых — 100 человек. Тогда результаты могут быть следующими:

В общей группе (10 молодых и 100 пожилых) из 110 человек:
* Лекарство A принимали 55 человек, и оно помогло 40 из них (40/55 = 72.7%).
* Лекарство B принимали 55 человек, и оно помогло 45 из них (45/55 = 81.8%).

Теперь при объединении данных оказывается, что лекарство B помогает большему проценту пациентов, чем лекарство A!

**Как возникает парадокс?**

Парадокс Симпсона возникает из-за скрытых переменных, которые влияют на анализ данных. В нашем примере такая скрытая переменная — это возраст пациентов. Объединяя данные, мы теряем важную информацию о том, как возраст влияет на эффективность лекарств. Этот эффект может привести к противоположным выводам в зависимости от того, рассматриваем ли мы данные в разбивке по группам или в совокупности. Подобный феномен возникает в реальном мире с определенной регулярностью. Главным моментом является то, что корреляция измеряет связь между двумя переменными <u>при прочих равных условиях</u>.

**Почему это важно?**

Парадокс Симпсона показывает, как важно учитывать все переменные при анализе данных и не делать поспешных выводов, основываясь только на совокупных данных. Он подчеркивает необходимость внимательного подхода к дизайну экспериментов и интерпретации их результатов.

Единственный реальный способ избежать таких неприятностей - это знать свои данные и обеспечивать проверку всех возможных спутывающих факторов. Очевидно, это не всегда возможно.

## Есть ли связь между NaN-значениями?

В данных практически всегда есть пропуски. Однако есть разница в их происхождении — так, иногда и из пропусков можно выделить полезную информацию о данных, например, если пропуски появились не случайно.

1. **Пропуски, независимые от данных (Missing Completely at Random, MCAR)**:
Пропуски данных считаются MCAR, если вероятность появления пропуска не зависит ни от значений самой переменной с пропусками, ни от других переменных. Такие пропуски происходят случайным образом, и причины отсутствия данных не связаны с данными, которые мы анализируем.

Пример:
Предположим, что в процессе сбора данных произошел сбой в оборудовании, и часть данных была случайно потеряна. В таком случае пропуски являются полностью случайными.

2. **Пропуски, зависящие от других переменных (Missing at Random, MAR)**:
Пропуски данных считаются MAR, если вероятность появления пропуска зависит от значений других переменных в наборе данных, но не от значений самой переменной с пропусками. То есть, если есть информация из других колонок, можно частично объяснить причину пропусков.

Пример:
Допустим, в медицинском наборе данных есть переменные "Возраст" и "Доход". Пропуски в переменной "Доход" могут чаще встречаться у людей младшего возраста, потому что они могут быть студентами или еще не начали зарабатывать деньги. В данном случае пропуски в доходах зависят от возраста.

Или другой пример: пропуск сразу в нескольких столбцах, и такие пропуски образуют паттерн. Такое может быть, например, в клиентских данных сетей, когда скидочная карта клиента утеряна в базе, и напротив статуса карты, имени владельца и прочей информации стоит NaN.

3. **Пропуски, зависящие от самой переменной (или от невыявленной переменной) (Missing Not at Random, MNAR)**:
Пропуски данных называются MNAR, если вероятность того, что данные будут отсутствовать, связана с самой переменной, содержащей пропуски. Это одна из самых сложных ситуаций, потому что информация о пропуске прямо зависит от недостающих данных.

Пример:
Представим, что в анкете о доходах люди с высоким доходом реже отвечают на вопрос о своем заработке, потому что они не хотят раскрывать эту информацию. В данном случае факт пропуска напрямую связан с уровнем дохода.

## Выявление аномалий с помощью box-plot

**Выбросы, или аномальные значения** — это аномальные или экстремальные значения в данных, которые могут искажать результаты анализа или моделирования. Например, мы берем данные по росту населения и внезапно видим число 313. Очевидно, что человек такого роста быть не может, и такое значение создает большой разброс в данных. К тому же это значение одно или в очень малом количестве на фоне общей выборки, которое могло возникнуть в результате ошибки датчика, повреждения файла и другое, и от него желательно каким-либо образом избавиться.

Ящик с усами представляет собой графическое отображение пятичисленного резюме данных (минимум, первый квартиль, медиана, третий квартиль, максимум). Этот вид графика активно применяется для выявления выбросов в данных на основе **метода межквартильного размаха**.

Межквартильный размах $IQR$ рассчитывается как разница между третьим и первым квартилями:
$$ IQR=Q3−Q1 $$

* Центральный ящик boxplot строится от первого квартиля (Q1) до третьего квартиля (Q3), и его длина равна IQR. Это основная часть boxplot, которая показывает средние 50% данных.
* Медиана (Q2) отображается линией внутри ящика, представляя срединное значение набора данных.
* "Усы" обычно простираются до $Q1−k×IQR$ и $Q3+k×IQR$, где обычно множитель $k = 1.5$. Эти значения служат границами для определения потенциальных выбросов.
* Выбросы (outliers) - данные, лежащие за пределами "усов", могут быть отдельно обозначены на графике.

## Поиск аномалий с помощью других методов

### Isolation Forest

— это метод для обнаружения аномалий, который основывается на идее, что аномалии легче изолировать, чем нормальные наблюдения. Он строит случайные "деревья", разделяя данные, и оценивает, насколько легко каждое значение "изолировать". Чем быстрее значение изолируется, тем больше вероятность, что оно выброс.

* Шаг 1: Метод строит несколько случайных деревьев, в которых выборки разбиваются на подмножества по случайным признакам и значениям. Каждый раз, когда мы выбираем признак и значение, мы делим данные на две части.

* Шаг 2: Аномальные точки данных требуют меньшего количества разбиений для изоляции, потому что они расположены далеко от остальной массы данных. Таким образом, количество разбиений, необходимых для изоляции точки, указывает на её аномальность.

* Шаг 3: После того как все деревья построены, алгоритм подсчитывает среднюю глубину изоляции каждой точки. Чем меньше глубина, тем более вероятно, что точка является аномалией.

### LOF (Local Outlier Factor)

Алгоритм оценивает выбросы на основе плотности данных. Он сравнивает плотность каждой точки с плотностью её соседей. Если плотность значительно ниже, точка считается выбросом.

Краткий алгоритм:
1. LOF вычисляет плотность окружения каждой точки, используя расстояние до k-ближайших соседей. Это позволяет оценить, насколько густо населена точка в сравнении с её соседями.
2. LOF сравнивает плотность точки с плотностями её соседей. Если точка имеет значительно меньшую плотность, чем её соседи, она считается аномалией.
3. LOF для каждой точки рассчитывается как отношение её плотности к плотности её соседей, то есть вопрос не в том, насколько изолирован образец, а в том, насколько он изолирован по отношению к окружающему району. Значения LOF больше 1 указывают на аномалии, чем больше значение, тем более аномальной считается точка.

# Чистка данных

## Стратегии ликвидации пропусков и выбросов

* **Удаление**

  * Удаление столбца
  * Удаление строки

Разберем подробнее удаление. Это самая жадная практика, и пользоваться ей надо осторожно. Если признак имеет большое количество пропусков и фактически не несет в себе полезной информации, можно удалить этот столбец. Если пропуски встречаются редко, можно удалить строки с пропусками.

* **Заполнение (импутация)**

  1. Мода, медиана, среднее — заполнить пропуски средним значением (для числовых данных), медианой или модой, самый простой подход.

  2. Предыдущее значение — больше подходит для временных рядов (дат, например).

  3. Внутригрупповое значение — подходит, если в данных имеются явные группы (например, гендерное разделение, для которых данные отличаются) или сезоны (например, продажи для лета и зимы отличаются).

  4. Множественное восстановление — прогнозирует пропущенные значения на основе доступных значений переменной, в которой мы заполняем пропуски, и других переменных. Основная идея состоит в том, чтобы не просто "угадать" одно значение для пропуска, а создать несколько наборов данных с разными заполненными значениями, провести анализ на каждом из этих наборов и затем усреднить результаты.

  5. Линейная интерполяция — недостающие значения вычисляются на основе предположения, что изменения между соседними известными точками линейны. Проще говоря, метод предполагает, что между двумя точками данные изменяются равномерно, и вычисляет пропущенные значения как точки на этой линии.

  6. Создание нового признака для NaN — Nan-значение заменяется на какое-то другое, а рядом создается дополнительный столбец, куда вносится информация о том, были ли в этой строке для этого признака пропуск.



# Feature Engineering

Feature Engineering охватывает несколько процессов:

1. **Feature Creation, или создание признаков.**

Здесь привлекаются Ваши или знания эксперта доменной области для создания каких-то новых признаков на основе имеющихся данных. В том числе сюда можно отнести биннинг, разбиение даты на день-месяц-год. Биннинг — это разбиение непрерывного диапазона значений на "корзины". Например, имеется колонка "возраст", и после биннинга значение из этой колонки будет не сам возраст, а категория, например, 0-18, 18-25, 26-34 и т.д.

2. **Feature Transformation.**

На этом шаге идут всевозможные преобразования данных:
- кодирование категориальных переменных
- преобразование распределения
- масштабирование признаков


3. **Feature Selection, или отбор признаков.**

На этом этапе применяются различные методы для того, чтобы определить наиболее важные признаки во всем датасете.

Отдельно еще можно выделить две опциональные задачи:
- **балансировка данных**
- **сокращение размерности данных**

## Кодирование признаков

Кодирование категориальных признаков — это процесс преобразования категориальных данных (текстовых, номинальных или порядковых значений) в числовые, чтобы их можно было использовать в моделях машинного обучения. Большинство алгоритмов машинного обучения требуют числового представления признаков, поэтому категориальные данные нужно трансформировать.

## Масштабирование признаков (Feature Scaling)

В некоторых алгоритмах машинного обучения различия в масштабах признаков могут приводить к проблемам. Признаки с более высокими значениями могут оказывать непропорциональное влияние на итоговую модель.

Основная задача feature scaling — привести признаки к единой шкале, чтобы избежать диспропорций между значениями различных признаков. Например, в задаче предсказания цены дома, площадь может измеряться в тысячах квадратных метров, а количество комнат — в единицах. Если эти признаки напрямую использовать в модели, то она может считать, что площадь вносит больший вклад просто из-за масштабов величин. Кроме того, при работе с алгоритмами, использующими градиентный спуск, о котором будем говорить дальше в курсе, различие в масштабах может замедлить процесс сходимости из-за особенностей вычислений в алгоритмах. Условно визуализировать это можно следующей картинкой — из-за разности в масштабах поверхность, у которой нужно найти минимум, имеет форму вытянутой искаженной поверхности, и дойти до минимума становится очень тяжело или невозможно.

Некоторые алгоритмы не зависят от масштаба признаков. Это касается деревьев решений (например, Random Forest), так как они разделяют данные на основе условий, а не расстояний.

### Минимаксная нормализация

1. Нормализация (Min-Max Scaling)

Нормализация — это метод, при котором значения признаков преобразуются в диапазон от 0 до 1 (или другой диапазон).

$$X_{normalized} = \frac {X−X_{min}} {X_{max} −X_{min}}  $$
​

где
$X$ — исходное значение признака,
$X_{min}$ и $X_{max}$ — минимальное и максимальное значения признака соответственно.

### Стандартизация

Стандартизация (Z-score Normalization или Standard Scaling)

Стандартизация преобразует данные так, что среднее значение признака становится 0, а стандартное отклонение — 1.

$$ X_{scaled} = \frac {X-\mu} {\sigma} $$
где $X$ — исходное значение признака, $\mu$ — среднее значение признака, $\sigma$ — стандартное отклонение признака.

# Дисбаланс классов

Дисбаланс классов (class imbalance) в данных — это ситуация, когда количество примеров в одном классе значительно превышает количество примеров в другом (или других) классах. Например, в задаче обнаружения мошенничества, где мошеннические транзакции могут составлять всего 10% от общего числа транзакций, класс "мошенничество" будет сильно недопредставлен по сравнению с классом "не мошенничество".

Модели машинного обучения могут быть склонны предсказывать преобладающий класс, чтобы минимизировать ошибки на обучающем наборе данных, что приводит к низкой точности предсказаний для редких классов. Например, модель, которая всегда предсказывает преобладающий класс, может иметь высокую точность, но при этом быть бесполезной для задачи классификации редких событий.

Для борьбы с этим явлением есть несколько подходов:

* Увеличение выборки (Oversampling): Увеличение количества примеров редких классов.
* Уменьшение выборки (Undersampling): Уменьшение количества примеров преобладающего класса. Это может привести к потере информации, но может быть эффективным, если данных достаточно много.
* Взвешивание классов: При обучении модели можно назначить разные веса для классов, чтобы модель учитывала редкие классы с большим весом.

# Сокращение размерности данных и проклятие размерности

**Проклятие размерности** — это термин, описывающий проблемы, которые возникают при работе с данными в высокоразмерных пространствах. В машинном обучении и статистике размерность обычно означает количество признаков или переменных в наборе данных. Как правило, по мере увеличения количества признаков модель сталкивается с рядом сложностей, которые ухудшают её производительность и требуют больше ресурсов для вычислений.

В высокоразмерных пространствах данные становятся сильно разреженными. Это означает, что с увеличением количества признаков каждая точка данных находится всё дальше от других. При малом числе данных многие точки в пространстве могут оказаться "изолированными", что затрудняет поиск значимых паттернов.

По мере увеличения количества признаков количество данных, необходимых для адекватного обучения моделей, растёт экспоненциально. Это связано с тем, что каждая новая переменная добавляет дополнительную степень свободы, требующую больше данных для корректной оценки.

**Сокращение размерности** (или dimensionality reduction) — это процесс уменьшения числа признаков (измерений) в данных, который помогает упростить анализ и обучение моделей машинного обучения, при этом стараясь минимизировать потерю важной информации. Это один из способов борьбы с проклятием размерности. Также есть и другое применение — человеку сложно визуализировать данные в пространстве с высокой размерностью (например, больше 3 признаков). Сокращение размерности позволяет свести данные к 2D или 3D пространствам, что делает визуализацию возможной. Однако стоит помнить, что полученное отображение нельзя как-то интерпретировать — до этого в датасете были колонки с конкретными признаками и их значением, то после применения метода сокращения размерности трудно сказать, какие исходные признаки повлияли на расположение точек в проекции. Эти методы хорошо подходят для визуализации кластеров или группировки объектов, но плохо для интерпретации признаков.

Методы линейного сокращения размерности:
* Метод главных компонент (Principal Component Analysis, PCA): Один из самых популярных методов. Он преобразует исходные данные в новое пространство меньшей размерности, где новые переменные (главные компоненты) являются линейными комбинациями исходных признаков, и ранжируются по тому, сколько дисперсии они объясняют.

Методы нелинейного сокращения размерности:
* t-SNE: Этот метод проецирует данные в пространство меньшей размерности (обычно двумерное), сохраняя их локальные структуры. Часто используется для визуализации.
* UMAP: Унифицированное аппроксимирование и проекция для измерения, аналог t-SNE, но с лучшей производительностью на больших наборах данных.
