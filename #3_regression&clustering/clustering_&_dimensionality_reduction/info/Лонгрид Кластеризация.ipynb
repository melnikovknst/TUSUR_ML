# Постановка задачи

**Кластеризация** — это объединение объектов или наблюдений в группы, называемые кластерами, на основе близости значений их признаков.

В результате в каждом кластере будут находиться объекты:
1. похожие по своим свойствам друг на друга
2. отличающиеся от объектов, которые содержатся в других кластерах

Чем больше подобие объектов внутри кластера и чем сильнее их отличие от объектов в других кластерах, тем лучше кластеризация.

## Теорема Клейнберга о невозможности

Одно из ограничений в задаче кластеризации – это теорема о невозможности, которую сформулировал Клейнберг. Если говорить проще, то нет идеального алгоритма кластеризации. Даже лучшие алгоритмы кластеризации могут быть непригодными для определенных типов данных или ситуаций.

## Ограничения кластеризации

Помимо этого, у кластеризации есть дополнительно ряд ограничений. Задача кластеризации имеет субъективное решение, то есть нет единого и однозначного способа разбить данные на кластеры.

Причины:

* Не существует однозначного критерия качества кластеризации. Для определения качества кластеризации и оценки выделенных кластеров необходим эксперт предметной области;
* Число кластеров, как правило, заранее не известно и выбирается по субъективным критериям. Даже если алгоритм не требует изначального знания о числе классов, конкретные реализации зачастую требуют указать этот параметр. Об этом поговорим дальше в лекции;
Результат кластеризации существенно зависит от метрики. Однако существует ряд рекомендаций по выбору метрик для определенных классов задач.
* Формализация задачи кластеризации. Насколько жесткие рамки поставите – настолько и можно приблизиться к наилучшему результату.
* Кластеризация может давать несколько хороших, но различных решений, поэтому решение надо подбирать в зависимости от вашей задачи.
* Данные по текущему набору признаков тяжело кластеризовать достаточно хорошо. И тогда возникает вопрос: решение есть, но оно не очень хорошее. Может, тогда решения нет вовсе?


## Индивидуальные особенности задачи

У каждой задачи кластеризации есть свои особенности:
1. формы кластеров, так как они могут формироваться в круг, ленты или очень странные, но в различимые человеческому глазу формы, которые могут тяжело даваться алгоритмам
2. учитывается ли в задаче вложенность кластеров
3. жесткая или мягкая эта кластеризация
4. является ли кластеризация основной задачей или результаты кластеризации будут использоваться дальше.

## Цели кластеризации

Кластеризация позволяет добиться следующих целей:
* улучшает понимание данных за счет выявления структурных групп;
* разбиение набора данных на группы схожих объектов позволяет упростить дальнейшую обработку и принятие решений, применяя к каждому кластеру свой метод анализа;
* позволяет компактно представлять и хранить данные. Для этого вместо хранения всех данных можно оставить по одному типичному наблюдению из каждого кластера;
* поиск новизны — обнаружение нетипичных объектов, которые не попали ни в один кластер.


## Области применения

Кластеризация используется во многих областях, помимо самой информатики. Очень заметная отрасль – маркетинг, интернет-магазины, где для формирования рекламного предложения необходимо сегментировать клиентские группы по товарам, активностям, данным клиентов и так далее. Еще в качестве примера – формирование тематических групп из наборов слов, группировка результатов поиска веб-сайтов и т.д.


# Основные методы кластеризации

1. Вероятностные алгоритмы - каждый объект множества принадлежит к одному из кластеров. К ним относятся K-Means, EM-алгоритм, FOREL и другие.

2. На основе нейронных сетей: С-means, сеть Кохонена, генетический алгоритм.

3. Иерархические алгоритмы -  у кластеров есть иерархия (вложенность). В таких алгоритмах один большой кластер разбивается на подмножества, а эти подмножества еще на подмножества, или наоборот: подмножества кластеров постепенно объединяются в один кластер.

4. Алгоритмы, которые сложно отнести к какой-то из предыдущих. К ним относится DBSCAN, KRAB.



# Способы найти начальное количество кластеров

Для некоторых алгоритмов кластеризации необходимо знать оптимальное или начальное (приблизительное) количество кластеров, для которого алгоритм может выполнять кластеризацию:
* метод Элбоу (метод локтя)
* индекс силуэта
* индекс Калински-Харабаза
* gap-statistics.


## Метод локтя

**Метод Элбоу (метод локтя)** - простой, но в то же время достаточно неточный и эвристический, и используется при кластеризации с помощью K-Means.

1. Выбрать какое-то число K, где вы предполагаете, что будет оптимальное значение.
2. Кластеризовать данные несколько раз, и в каждой следующей итерации количество кластеров увеличивается вплоть до K.
3. Полученные кластеры исследуются: оценивается метрика - сумма квадратов расстояний между каждой точкой и центроидом внутри каждого кластера, к которому она относится.

Если построим график зависимости этой метрики от количества кластеров, то в какой-то момент метрика начнет резко падать, так как кластеры начнут кучковаться, и этот **перегиб и есть оптимальное количество кластеров**. Однако такой перегиб может быть достаточно гладким, и тогда нельзя оценить четкое количество кластеров. Однако можно увидеть диапазон, где начинается и заканчивается перегиб, и этот диапазон количества кластеров использовать при расчете метода силуэта.

## Метод силуэта

**Коэффициент силуэта** позволяет оценить качество кластеризации,  а также выбрать начальное число кластеров.
1. Провести кластеризацию, например, через K-Means, и явно указать количество кластеров, на которые нужно кластеризовать данные.
2. Вычисление коэффициентов, которые присваиваются каждому объекту в кластере и образуют так называемый силуэт кластера. Коэффициенты изменяются от -1 до 1.
> Значения, близкие к 1, указывают на то, что объект является похожим на другие объекты в кластере и не похожим на объекты из других кластеров.
> Если большинство объектов имеют значения коэффициентов близкими к 1, можно утверждать, что кластерная структура хорошо выражена.
3. Вычисляется среднее значение коэффициента для этого предполагаемого числа кластеров. 4. 4. Повторять несколько раз для разного количества кластеров.

Оптимальное количество кластеров выбирается так, чтобы средний коэффициент силуэта был максимальным.

## Индекс Калинского-Харабаза

1. Провести кластеризацию, например, через K-Means.
2. Для каждого количества кластеров считается индекс Калинского-Харабаза и строится график зависимости индекса от количества кластеров.

 Индекс Калинского-Харабаза принимает значения от 0 до ∞. Чем больше значение индекса, тем лучше кластеризация.

# Основные алгоритмы кластеризации

## K-Means

Алгоритм K-Means очень популярен из-за скорости работы и приемлемых результатов для несложных задач.

Работает алгоритм следующим образом:
1. Выбирается число кластеров $k$;
2. Из исходного множества данных случайным образом выбираются $k_0$ наблюдений, которые будут служить начальными центрами кластеров.
3. Для каждого наблюдения $g$ исходного множества определяется ближайший к нему центр кластера $k_0$ (расстояния измеряются в заданной метрике). При этом записи, «притянутые» определенным центром, образуют начальные кластеры $r$.
4. Вычисляются центроиды $k$ — центры тяжести кластеров. Каждый **центроид** — это вектор, элементы которого представляют собой средние значения соответствующих признаков, вычисленные по всем записям кластера.
5. Центр кластера $r$ смещается в его центроид $k$, после чего центроид становится центром нового кластера r.
6. 3-й, 4-й и 5-ый шаги итеративно повторяются. Очевидно, что на каждой итерации происходит изменение границ кластеров и смещение их центров. В результате минимизируется расстояние между элементами внутри кластеров и увеличиваются междукластерные расстояния.

Варианты метрик расстояния:
* манхэттенское
* евклидово
* косинусная мера

Особенности алгоритма:
* нужно знать начальное количество кластеров $k$
* чувствительность к начальным значениям кластеров
* чувствительность к форме кластеров (лучше работает со сферообразными)
* плохо реагирует на шумы

### Способы подсчета расстояния

Подсчитывать расстояние между кластерами можно разными способами:

1. Евклидово расстояние:
$$ p(x, x') = \sqrt{\sum_{i=1}^{n} (x_i - x'_i)^2}
$$
Квадрат евклидова расстояния придает больше веса отдаленным друг от друга элементам в кластере
2. Манхэттенское расстояние $p(x, x') = \sum_{i=1}^{n} |x_i - x'_i|
$ — при его использовании влияние отдельных выбросов меньше сказывается на результате
3. Косинусная мера

## DBSCAN

Алгоритм **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) — это алгоритм кластеризации, основанной на плотности.
Идея алгоритма DBSCAN заключается в том, что кластеры образуют плотные области данных, которые отделены друг от друга относительно пустыми областями. Алгоритм DBSCAN определяет точки, расположенные в "густонаселенных" областях пространства характеристик, когда многие точки данных расположены близко друг к другу.

Алгоритм DBSCAN работает следующим образом:
1. Для каждой точки в наборе данных определяется ее окрестность. Окрестность точки - это набор всех точек, которые находятся на расстоянии не более $ϵ$ от нее.
2. Точки, у которых количество соседей в окрестности не менее $minPts$, называются основными точками.
3. Для каждой следующей соседней точки проверяется условие, находятся ли они в радиусе $\epsilon$, и в этом случае они называются свящными компонентами.
4. Связные компоненты основных точек образуют кластеры.
5. Все остальные точки называются выбросами.

Особенности алгоритма:
* сам определяет начальное количество кластеров с помощью гиперпараметров
* справляется даже с достаточно сложными формами кластеров
* значение "-1" зарезервировано для шумов
* более долгое вычисление
* гиперпараметры задаются вручную

## Иерархическая кластеризация

В иерархической кластеризации кластеризация может быть снизу-вверх (агломеративной), то есть от нескольких кластеров к единому, и наоборот, сверху-вниз (дивизионной), дробя исходный кластер на несколько. Для этих методов указывается linkage – мера расстояния между кластерами.

Алгоритм для агломеративной кластеризации:
1. Инициализация множества кластеров, каждая точка считается своим кластером. То есть для выборки размера N на первой итерации будет N кластеров.
> Также входным параметром алгоритму подается метрика расстояния между двумя кластерами. Самой популярной метрикой является расстояние Уорда.
2. На каждой итерации объединяется два кластера в один. Объединяющиеся кластера выбираются в соответствии с выбранным linkage. То есть в соответствии с выбранным нами расстоянием эти два кластера будут наиболее похожи и поэтому объединяются
3. Предыдущий шаг повторяется вплоть до объединения всех точек один кластер.

Для дивизионной кластеризации алгоритм выполняется наоборот, а начальное количество кластеров равно единице.

### Виды linkage

* одиночная связь — наименьшее расстояние между двумя элементами из разных кластеров
* полная связь — наибольшее расстояние между двумя элементами из разных кластеров
* средняя связь — среднее расстояние между двумя элементами из разных кластеров
* метод Уорда

# Метрики оценки качества кластеризации

* **Внутренние метрики** — если у данных не было исходной разметки данных на истинные классы, основанные на отделимости кластеров друг от друга.
* **Внешние метрики** — если есть разметка и поставлена задача кластеризации. Вычисляется как отделимость кластеров, так и их состав на основе существующей разметки.

> Метрик для обоих классов достаточно, и их выбор достаточно субъективен. Также выбор метрик зависит от области, для которой проводится кластеризация, так что говорить о каких-то универсальных метриках не приходится.



## Внутренние метрики

* среднее внутрикластерное расстояние
* среднее межкластерное расстояние
* индекс силуэта
* индекса Данна
* индекс Дэвиса-Болдина
и многие другие

**Внутрикластерное расстояние** математически записывается как мера кучности кластеров, и можно считать как между парами объектов, так и расстояние до центра:
$$ F_0 = \frac{\sum_{i < j} [y_i = y_j] \rho(x_i, x_j)}{\sum_{i < j} [y_i = y_j]} \rightarrow \min
 $$

**Междукластерное расстояни**е записывается как отделимость кластеров, и чем больше это расстояние, тем лучше:

$$ F_1 = \frac{\sum_{i < j} [y_i \neq y_j] \rho(x_i, x_j)}{\sum_{i < j} [y_i \neq y_j]} \rightarrow \max
 $$

**Индекс силуэта** измеряет, насколько хорошо объект "вписывается" в свой кластер по сравнению с его ближайшим соседним кластером. Эта метрика оценивает "компактность" кластера (насколько объекты внутри кластера похожи) и "разделимость" (насколько далеко объекты одного кластера от объектов других кластеров).

$$ S(i) = \frac{b_i - a_i}{max(a_i, b_i)}$$

Для объекта 𝑖 вычисляется два ключевых параметра:
* $a_i$ — среднее расстояние между объектом 𝑖 и всеми другими объектами в его кластере. Это называется внутрикластерным расстоянием и отражает, насколько объект близок к своему кластеру.

* $b_i$ — минимальное среднее расстояние между объектом 𝑖 и объектами из любого другого кластера. Это межкластерное расстояние, которое оценивает, насколько объект удален от ближайшего кластера.
> Если $S(i)$ близко к 1, это означает, что объект хорошо "вписывается" в свой кластер.
> Если $S(i)$ близко к 0, объект находится на границе между кластерами.
Отрицательные значения $S(i)$ указывают, что объект, вероятно, неправильно классифицирован и ближе к другому кластеру.


**Индекс Калинского-Харабаза** оценивает, насколько хорошо кластеры отделены друг от друга, и как сильно объекты внутри одного кластера сжаты (компактность).

$$CH = \frac{дисперсия_{межкластерная}}{дисперсия_{внутрикластерная}} * \frac{n-k}{k-1}$$
$n$ — общее количество точек данных,
$k$ — количество кластеров.

**Межкластерная дисперсия (между центроидами)** — это сумма квадратов расстояний между центроидами каждого кластера и глобальным центроидом (средним всех объектов). Она показывает, насколько далеко кластеры находятся друг от друга.

**Внутрикластерная дисперсия** — это сумма квадратов расстояний между объектами и их центроидом в рамках одного кластера. Она отражает компактность кластеров.
Значение индекса растет, когда межкластерная дисперсия увеличивается, а внутрикластерная дисперсия уменьшается. Чем выше индекс, тем лучше.


**Индекс Дэвиса-Болдина** измеряет среднее отношение внутрикластерного расстояния к межкластерному расстоянию. Чем ближе объекты одного кластера друг к другу и чем дальше кластеры находятся друг от друга, тем лучше кластеризация. Чем меньше значение индекса Дэвиса-Болдина, тем лучше. Это указывает на то, что кластеры хорошо разделены и компактны.

$$ DB = \frac{1}{k} \sum_{i=1}^{k} \max_{j \neq i} \left( \frac{\sigma_i + \sigma_j}{d_{ij}} \right) $$
В формуле $σ_i​$ — среднее расстояние между объектами кластера $i$ и его центроидом (т.е. внутрикластерная дисперсия),
$d_{ij}$ — расстояние между центроидами кластеров $i$ и $j$.


## Внешние метрики

* полнота
* гомогенность
* V-мера
* Индекс Рэнда (Rand Index)
* ARI (Adjusted Rand Index)
* AMI (Adjusted mutual information)
и многие другие

Если у датасета есть разметка, то можно использовать **полноту**, **гомогенность** и **V-меру**. **Гомогенность** (однородность) – кластер состоит только из объектов одного класса. **Полнота** – кластер состоит из всех объектов класса. И **V-мера** – среднее гармоническое между ними.

Для объяснения ARI нужно понять принцип **индекса Ранда (RI)**. **RI** измеряет долю согласованных пар объектов — это те пары объектов, которые либо находятся в одном кластере в обеих кластеризациях, либо в разных кластерах в обеих кластеризациях.
$$ RI = \frac{N_{rc}}{N_{total}}$$
$N_{rc}$ — число правильно кластеризованных пар, $N_{total}$ — общее число пар.

Чем ближе этот индекс к 1, тем лучше кластеризация согласована с истинной разметкой.


Проблема с RI заключается в том, что даже случайная кластеризация может давать высокие значения, особенно если много объектов. **ARI** решает эту проблему, вводя поправку на случайность.

$$ ARI = \frac{RI-Expected RI}{max(RI) -Expected RI}$$

Expected [RI] — математическое ожидание RI при случайной кластеризации.
Значения ARI интерпретируются так:
* 1: Полное соответствие между двумя кластеризациями.
* 0: Случайное распределение кластеров.
* Отрицательные значения: Кластеризация хуже, чем случайная.


ARI используется для сравнения результата кластеризации с известной разметкой данных (например, с реальными классами в задачах классификации или сегментации).



# Принцип работы PCA

**PCA** (Principal Component Analysis) — это метод уменьшения размерности данных, который используется для преобразования большого числа коррелированных переменных в меньшее количество некоррелированных переменных, называемых главными компонентами. Основная идея PCA — это найти такие направления (компоненты) в данных, которые захватывают наибольшую часть дисперсии (вариации) в этих данных, при этом уменьшая размерность. **Главные компоненты** - это новые переменные, с помощью которых мы и будем описывать наши объекты.

Идея состоит в том, что 10-мерные данные дают 10 главных компонент, но **PCA пытается поместить максимум возможной информации в первый компонент, затем максимум оставшейся информации во второй** и так далее.

Алгоритм работы:
1. Центрирование данных

Прежде чем начать процесс PCA, необходимо **центрировать** данные. Это означает, что для каждого признака (столбца) нужно вычесть его среднее значение. Центрирование помогает сделать данные симметричными относительно начала координат, что важно для корректной работы алгоритма PCA.

2. Вычисление ковариационной матрицы

После центрирования данных нужно вычислить **ковариационную матрицу**, которая измеряет, насколько признаки коррелируют друг с другом. Задача PCA — найти такие направления, вдоль которых ковариация между признаками будет минимальной (или нулевая), а дисперсия будет максимальной.

3. Поиск собственных векторов и собственных значений

Теперь нужно найти собственные векторы и собственные значения ковариационной матрицы. Это ключевая часть алгоритма PCA, так как собственные векторы указывают направления (оси), вдоль которых данные имеют наибольшую вариацию, а собственные значения показывают, насколько велика эта вариация. Когда мы ищем собственные вектора ковариационной матрицы, это находит такие направления в пространстве данных, вдоль которых дисперсия (разброс данных) максимальна. То есть собственные вектора показывают основные направления, вдоль которых данные "развёрнуты" сильнее всего, когда как собственное число показывает величину дисперсии вдоль этого направления.
* **Собственные векторы** — это направления, вдоль которых будет происходить преобразование данных.
* **Собственные значения** — это величины, которые показывают, сколько дисперсии захватывает каждая из этих осей.

4. Выбор главных компонент

После нахождения собственных векторов и собственных значений их упорядочивают по убыванию собственных значений. Первые несколько векторов с наибольшими собственными значениями называются **главными компонентами**, так как они захватывают наибольшую долю вариации в данных.

Если собственные значения упорядочены, то первая главная компонента — это направление с максимальной вариацией в данных, вторая главная компонента — это направление, перпендикулярное первой, с наибольшей оставшейся вариацией, и так далее.

5. Преобразование данных

Теперь можно проецировать исходные данные на новые оси (собственные векторы). Это даёт новое представление данных с меньшим числом признаков. Матрица преобразования состоит из первых k собственных векторов.

Допустим, у нас есть набор данных с тремя признаками $x_1$, $x_2$, $x_3$​, которые сильно коррелируют друг с другом. Применяя PCA, мы мож1ем преобразовать эти три признака в две главные компоненты, которые захватывают 90% дисперсии исходных данных. В результате получим новый набор данных с двумя некоррелированными признаками (компонентами), которые лучше отражают структуру данных. Однако стоит учитывать, что PCA дает линейное преобразование и не дает захвата сложных нелинейных связей, для которых уже используются T-SNE и UMAP.
