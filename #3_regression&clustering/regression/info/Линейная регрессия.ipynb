## –ú–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –∏ —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å

–î–ª—è –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å –∫–∞–∫ –æ–¥–∏–Ω –∏–∑ –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è **–º–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤** (Ordinary Least Squares, OLS). –û–Ω –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞–π—Ç–∏ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É—é—Ç –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—É—é –æ—à–∏–±–∫—É.

–ó–∞–¥–∞—á–∞ –º–µ—Ç–æ–¥–∞ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–∏ —Ç–∞–∫–∏—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ $w_0, w_1, ..., w_d$, –∫–æ—Ç–æ—Ä—ã–µ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É—é—Ç **—Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—à–∏–±–æ–∫** ($RSS$, risidual sum of squares, –∏–ª–∏ $SSE$ - sum of squared errors) ‚Äî —ç—Ç–æ –æ–¥–Ω–∞ –∏–∑ —Ñ—É–Ω–∫—Ü–∏–π –ø–æ—Ç–µ—Ä—å –¥–ª—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:

$$ RSS = \sum_{i=1}^N (y_i - \hat{y_i})^2 = \sum_{i=1}^N (y_i - (w_0 + w_1x_{i1} + w_2x_{i2} + ... + w_dx_{id}))^2$$

–≠—Ç—É —Ñ–æ—Ä–º—É–º—É –º–æ–∂–Ω–æ —Ç—Ä–∞–∫—Ç–æ–≤–∞—Ç—å —Ç–∞–∫: SSE - —ç—Ç–æ —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Ä–∞–∑–Ω–∏—Ü—ã –º–µ–∂–¥—É —ç—Ç–∞–ª–æ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏, –æ–±–æ–∑–Ω–∞—á–µ–Ω–Ω—ã—Ö –∫–∞–∫ $y_i$, –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π $\hat{y_i}$. –¢–æ –µ—Å—Ç—å –º—ã –≤—ã–±–∏—Ä–∞–µ–º —Å—Ç—Ä–æ–∫—É –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞ —Å –µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ —Ü–µ–ª–µ–≤—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º, –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ–¥—Å—Ç–∞–≤–ª—è–µ–º –≤ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –ü–µ—Ä–µ–º–Ω–æ–∂–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∏ –±—É–¥–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ–º. –ü–æ–ª—É—á–µ–Ω–Ω–æ–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –≤—ã—á–∏—Ç–∞–µ–º –∏–∑ —Ü–µ–ª–µ–≤–æ–≥–æ –∏ –≤–æ–∑–≤–æ–¥–∏–º –≤ –∫–≤–∞–¥—Ä–∞—Ç, —ç—Ç–æ –∏ –±—É–¥–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç –æ—à–∏–±–∫–∏. –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ–µ –ø–æ–≤—Ç–æ—Ä—è–µ–º –¥–ª—è –≤—Å–µ—Ö —Å—Ç—Ä–æ–∫ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ, –∏ —Å—É–º–º–∏—Ä—É–µ–º –≤ –∏—Ç–æ–≥–æ–≤—É—é –æ—à–∏–±–∫—É. –¶–µ–ª—å - –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å —ç—Ç—É –∏—Ç–æ–≥–æ–≤—É—é –æ—à–∏–±–∫—É.

–ú–æ–∂–µ—Ç –≤–æ–∑–Ω–∏–∫–Ω—É—Ç—å –≤–æ–ø—Ä–æ—Å: –∞ –∑–∞—á–µ–º –≤–æ–∑–≤–æ–¥–∏—Ç—å –≤ –∫–≤–∞–¥—Ä–∞—Ç? –ù–∞ —ç—Ç–æ –µ—Å—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –æ—Ç–≤–µ—Ç–æ–≤:

## –ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ OLS

–°—É—â–µ—Å—Ç–≤—É–µ—Ç –≥–æ—Ç–æ–≤–æ–µ –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ OLS, –∑–∞–ø–∏—Å–∞–Ω–Ω–æ–µ –≤ –º–∞—Ç—Ä–∏—á–Ω–æ–π —Ñ–æ—Ä–º–µ:

$$ X^TXw = X^Ty$$

–†–µ—à–µ–Ω–∏–µ —ç—Ç–∏—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π –¥–∞—ë—Ç:

$$ w = (X^TX)^{-1}X^Ty$$

$w$  ‚Äì –∏—Å–∫–æ–º—ã–µ –≤–µ—Å–∞, –∏–ª–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã, –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏

$X$ ‚Äì –º–∞—Ç—Ä–∏—Ü–∞ —Ñ–∏—á–µ–π

$X^T$ ‚Äì —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —Ñ–∏—á–µ–π

$y$  ‚Äì –≤–µ–∫—Ç–æ—Ä —Ü–µ–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π

$-1$  ‚Äì –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏–µ –æ–±—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã

–≠—Ç–æ —Ä–µ—à–µ–Ω–∏–µ –∏ –µ—Å—Ç—å –≤–µ–∫—Ç–æ—Ä –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –∫–æ—Ç–æ—Ä—ã–π –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –≥–∏–ø–µ—Ä–ø–ª–æ—Å–∫–æ—Å—Ç—å, –Ω–∞ –∫–æ—Ç–æ—Ä—É—é –ø—Ä–æ–µ—Ü–∏—Ä—É–µ—Ç—Å—è –≤–µ–∫—Ç–æ—Ä $y$.

**–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø–æ–¥—Ä–æ–±–Ω–µ–µ –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏–π –≤—ã–≤–æ–¥**

–ó–∞–¥–∞—á–∞ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å–æ—Å—Ç–æ–∏—Ç –≤ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–∏ —Ç–∞–∫–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ $w$, –∫–æ—Ç–æ—Ä—ã–π –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—É—é –æ—à–∏–±–∫—É –º–µ–∂–¥—É –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ $\hat{y} = Xw$ –∏ –Ω–∞–±–ª—é–¥–∞–µ–º—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ $y$.

–§—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å, –∫–æ—Ç–æ—Ä—É—é –º—ã –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ–º:

$$ L(w) = || y - Xw||^2 = (y - Xw)^T(y - Xw) $$

–ó–∞–ø–∏—Å—å $|| y - Xw||$ ‚Äî —ç—Ç–æ –µ–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ –æ—à–∏–±–æ–∫, –∫–æ—Ç–æ—Ä–∞—è –æ–±–æ–∑–Ω–∞—á–∞–µ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏ $y$ (–Ω–∞–±–ª—é–¥–∞–µ–º—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏) –∏ $Xw$ (–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏). –ù–∞—Å –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏—è –∫–≤–∞–¥—Ä–∞—Ç–∞ –Ω–æ—Ä–º—ã, —Ç–æ –µ—Å—Ç—å $|| y - Xw||^2$. –≠—Ç–æ —Ç–æ –∂–µ —Å–∞–º–æ–µ, —á—Ç–æ –∏ —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π (–æ—à–∏–±–æ–∫) –º–µ–∂–¥—É –Ω–∞–±–ª—é–¥–∞–µ–º—ã–º–∏ –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏:

$$ || y - Xw||^2 = \sum_{i=1}^n{(y_i - (Xw)_i)^2} $$

–°–º—ã—Å–ª —ç—Ç–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è ‚Äî —ç—Ç–æ –æ–±—â–∞—è –æ—à–∏–±–∫–∞ –º–æ–¥–µ–ª–∏, –∏–∑–º–µ—Ä—è–µ–º–∞—è –∫–∞–∫ —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Ä–∞–∑–Ω–æ—Å—Ç–µ–π –º–µ–∂–¥—É —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏–º–∏ –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏.

–¢–µ–ø–µ—Ä—å –ø–µ—Ä–µ–π–¥—ë–º –∫ –≤—ã—Ä–∞–∂–µ–Ω–∏—é $|| y - Xw||^2 = (y - Xw)^T(y - Xw)$. –≠—Ç–æ –∞–ª–≥–µ–±—Ä–∞–∏—á–µ—Å–∫–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –∫–≤–∞–¥—Ä–∞—Ç–∞ –Ω–æ—Ä–º—ã —á–µ—Ä–µ–∑ –º–∞—Ç—Ä–∏—á–Ω–æ–µ —É–º–Ω–æ–∂–µ–Ω–∏–µ:
* $(y - Xw)^T$ ‚Äî —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π (—Å—Ç—Ä–æ–∫–∞),
* $(y - Xw)$ ‚Äî  –≤–µ–∫—Ç–æ—Ä –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π (—Å—Ç–æ–ª–±–µ—Ü).

–£–º–Ω–æ–∂–µ–Ω–∏–µ —ç—Ç–∏—Ö –¥–≤—É—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –¥–∞—ë—Ç —Å–∫–∞–ª—è—Ä ‚Äî —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤–µ–∫—Ç–æ—Ä–∞ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π. –ò–º–µ–Ω–Ω–æ —ç—Ç–æ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –µ–≤–∫–ª–∏–¥–æ–≤–æ–π –Ω–æ—Ä–º–µ –≤ –∫–≤–∞–¥—Ä–∞—Ç–µ.

–î–ª—è –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ –æ—à–∏–±–∫–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å –ø–æ –∫–∞–∂–¥–æ–º—É –ø–∞—Ä–∞–º–µ—Ç—Ä—É $w$ –∏ –ø—Ä–∏—Ä–∞–≤–Ω—è—Ç—å –∏—Ö –∫ –Ω—É–ª—é. –î–ª—è —ç—Ç–æ–≥–æ:

0. –ü—Ä–µ–∂–¥–µ —á–µ–º –≤—ã—á–∏—Å–ª—è—Ç—å –≥—Ä–∞–¥–∏–µ–Ω—Ç, —Ä–∞—Å–∫—Ä–æ–µ–º —Å–∫–æ–±–∫–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Å–≤–æ–π—Å—Ç–≤ —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è –∏ —É–º–Ω–æ–∂–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü:

$$ L(w) = (y - Xw)^T(y - Xw) = y^Ty - 2w^TX^Ty + w^TX^TXw$$

1. –ó–∞–ø–∏—à–µ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å –ø–æ $w$:

$$ \frac{\partial L(w)}{\partial w} = -2X^Ty + 2x^TXw = -2X^T (y - Xw)$$

2. –ü—Ä–∏—Ä–∞–≤–Ω–∏–≤–∞–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç –∫ –Ω—É–ª—é –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º—É–º–∞:

$$-2X^T (y - Xw) = 0$$

3. –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ:

$$ X^Ty = X^TXw$$

4. –†–µ—à–∞–µ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ $w$:

$$ w = (X^TX)^{-1}X^Ty$$

–≠—Ç–æ —Ä–µ—à–µ–Ω–∏–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç, –µ—Å–ª–∏ –º–∞—Ç—Ä–∏—Ü–∞ $X^TX$ –æ–±—Ä–∞—Ç–∏–º–∞. –ï—Å–ª–∏ –º–∞—Ç—Ä–∏—Ü–∞ –Ω–µ–æ–±—Ä–∞—Ç–∏–º–∞ –∏–ª–∏ –ø–ª–æ—Ö–æ –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–∞ (–∫–æ–≥–¥–∞ –µ—ë –¥–µ—Ç–µ—Ä–º–∏–Ω–∞–Ω—Ç –±–ª–∏–∑–æ–∫ –∫ –Ω—É–ª—é), —Ç–æ –≤–æ–∑–Ω–∏–∫–∞—é—Ç –ø—Ä–æ–±–ª–µ–º—ã —Å –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ–º –æ–±—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã.

## –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –Ω–∞ –æ—Å–Ω–æ–≤–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ –∏ MSE

–¢–∞–∫–∂–µ –æ–¥–Ω–æ–π –∏–∑ –Ω–∞–∏–±–æ–ª–µ–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω—ë–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π –ø–æ—Ç–µ—Ä—å —è–≤–ª—è–µ—Ç—Å—è —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–∞—è –æ—à–∏–±–∫–∞ ($MSE$, Mean Squared Error). $MSE$ –∏–∑–º–µ—Ä—è–µ—Ç —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –æ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –º–æ–¥–µ–ª—å—é:

 $$MSE = \frac{1}{n} \sum_{i=1}^n{(y_i - \hat{y_i})^2}$$

–ö–∞–∫ –º–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, –º–µ–∂–¥—É RSS –∏ MSE –µ—Å—Ç—å —Å–≤—è–∑—å ‚Äî MSE —è–≤–ª—è–µ—Ç—Å—è —É—Å—Ä–µ–¥–Ω–µ–Ω–∏–µ–º RSS:

 $$MSE = \frac{1}{n} \sum_{i=1}^n{(y_i - \hat{y_i})^2} = \frac{1}{n}RSS$$

 –ü—Ä–æ—Ü–µ—Å—Å –æ–±—É—á–µ–Ω–∏—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å–≤–æ–¥–∏—Ç—Å—è –∫ –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ MSE, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ —Ç–∞–∫–∏—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ $w$, –∫–æ—Ç–æ—Ä—ã–µ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É—é—Ç —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫—É—é –æ—à–∏–±–∫—É. –ß–∞—Å—Ç—ã–π —Å–ø–æ—Å–æ–± –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ MSE ‚Äî –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏—Ç–µ—Ä–∞—Ç–∏–≤–Ω—ã–π –º–µ—Ç–æ–¥, –Ω–∞–ø—Ä–∏–º–µ—Ä, **–≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫**. –£ –≤—ã–±—Ä–∞–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ (–≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ MSE) –µ—Å—Ç—å —Å–≤–æ–π –º–∏–Ω–∏–º—É–º. –ò –Ω–∞—à–∞ —Ü–µ–ª—å ‚Äî —Å –ø–æ–º–æ—â—å—é –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ –¥–æ–±—Ä–∞—Ç—å—Å—è –¥–æ —ç—Ç–æ–≥–æ –º–∏–Ω–∏–º—É–º–∞.

 –í —Å–ª—É—á–∞–µ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫, —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –Ω–∞–∏–ª—É—á—à–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ª–∏–Ω–µ–π–Ω–æ–π –º–æ–¥–µ–ª–∏, –∫–æ—Ç–æ—Ä—ã–µ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É—é—Ç –æ—à–∏–±–∫–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è.  –ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫ –ø–æ–º–æ–≥–∞–µ—Ç –º–µ–¥–ª–µ–Ω–Ω–æ –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –º–æ–¥–µ–ª–∏ —Ç–∞–∫, —á—Ç–æ–±—ã MSE —É–º–µ–Ω—å—à–∞–ª—Å—è. –≠—Ç–æ –¥–µ–ª–∞–µ—Ç—Å—è —à–∞–≥ –∑–∞ —à–∞–≥–æ–º. –ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ, —á—Ç–æ MSE ‚Äî —ç—Ç–æ —Ä–µ–ª—å–µ—Ñ–Ω–∞—è –≥–æ—Ä–∫–∞, –∞ –Ω–∞—à–∞ —Ü–µ–ª—å ‚Äî –¥–æ–±—Ä–∞—Ç—å—Å—è –¥–æ –µ—ë –ø–æ–¥–Ω–æ–∂—å—è, –≥–¥–µ –æ—à–∏–±–∫–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–∞. **–ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫ –ø–æ–º–æ–≥–∞–µ—Ç –ø–æ–Ω—è—Ç—å, –≤ –∫–∞–∫–æ–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–∏ "—Å–ø—É—Å–∫–∞—Ç—å—Å—è"**, —Ç–æ –µ—Å—Ç—å —É–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫ –Ω—É–∂–Ω–æ –∏–∑–º–µ–Ω–∏—Ç—å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã, —á—Ç–æ–±—ã –æ—à–∏–±–∫–∞ —É–º–µ–Ω—å—à–∞–ª–∞—Å—å. –≠—Ç–æ—Ç –ø—Ä–æ—Ü–µ—Å—Å –ø—Ä–æ–¥–æ–ª–∂–∞–µ—Ç—Å—è: –º—ã –∏–∑–º–µ–Ω—è–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –º–æ–¥–µ–ª–∏, –ø–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞–µ–º MSE –∏ –ø—Ä–æ–¥–æ–ª–∂–∞–µ–º –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞—Ç—å, –ø–æ–∫–∞ –Ω–µ –Ω–∞–π–¥—ë–º —Ç–µ, –∫–æ—Ç–æ—Ä—ã–µ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É—é—Ç MSE.

## –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –º–æ–¥–µ–ª–∏

–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ—Å—Ç–∞. –û—Å–Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏ –≤—ã—Å—Ç—É–ø–∞—é—Ç –µ–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã $w$: –æ–Ω–∏ –æ—Ç—Ä–∞–∂–∞—é—Ç –≤–∫–ª–∞–¥ –∫–∞–∂–¥–æ–π –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –≤ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.

* $w_0$  ‚Äî —ç—Ç–æ —Å–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω –∏–ª–∏ –∏–Ω—Ç–µ—Ä—Å–µ–ø—Ç, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π $y$, –∫–æ–≥–¥–∞ –≤—Å–µ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ —Ä–∞–≤–Ω—ã –Ω—É–ª—é. –≠—Ç–æ—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø–æ–º–æ–≥–∞–µ—Ç —Å–º–µ—Å—Ç–∏—Ç—å –ª–∏–Ω–∏—é —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –≤–¥–æ–ª—å –æ—Å–∏ $y$, —á—Ç–æ–±—ã –ª—É—á—à–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–º.

* $w_1, w_2, ..., w_d$ ‚Äî —ç—Ç–æ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –Ω–∞–∫–ª–æ–Ω–∞, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç, –Ω–∞ —Å–∫–æ–ª—å–∫–æ –∏–∑–º–µ–Ω–∏—Ç—Å—è —Ü–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è $y$, –µ—Å–ª–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è –Ω–µ–∑–∞–≤–∏—Å–∏–º–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è $x_1, x_2, .., x_d$ –∏–∑–º–µ–Ω–∏—Ç—Å—è –Ω–∞ –µ–¥–∏–Ω–∏—Ü—É –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ –≤—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –æ—Å—Ç–∞—é—Ç—Å—è –Ω–µ–∏–∑–º–µ–Ω–Ω—ã–º–∏.

# –†–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è

–ö–æ–≥–¥–∞ –º–æ–¥–µ–ª—å —Å–ª–∏—à–∫–æ–º —Ö–æ—Ä–æ—à–æ –ø–æ–¥—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç—Å—è –ø–æ–¥ –æ–±—É—á–∞—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ, –æ–Ω–∞ –º–æ–∂–µ—Ç "–ø–µ—Ä–µ—É—á–∏—Ç—å—Å—è" (–ø–µ—Ä–µ–æ–±—É—á–∏—Ç—å—Å—è) –∏ –Ω–∞—á–∞—Ç—å –¥–µ–ª–∞—Ç—å –æ—à–∏–±–∫–∏ –Ω–∞ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –≠—Ç–æ –ø–æ—Ö–æ–∂–µ –Ω–∞ —Å–∏—Ç—É–∞—Ü–∏—é, –∫–æ–≥–¥–∞ –≤—ã —É—á–∏—Ç–µ—Å—å –±—Ä–æ—Å–∞—Ç—å –º—è—á –≤ –æ–¥–Ω—É –∏ —Ç—É –∂–µ –∫–æ—Ä–∑–∏–Ω—É —Å –æ–¥–Ω–æ–≥–æ –∏ —Ç–æ–≥–æ –∂–µ –º–µ—Å—Ç–∞, –Ω–æ –∫–∞–∫ —Ç–æ–ª—å–∫–æ —Ç–µ–±–µ –Ω—É–∂–Ω–æ –±—Ä–æ—Å–∏—Ç—å —Å –¥—Ä—É–≥–æ–≥–æ –º–µ—Å—Ç–∞ –∏–ª–∏ –≤ –¥—Ä—É–≥—É—é –∫–æ—Ä–∑–∏–Ω—É, —Å–ª–µ–¥—É–µ—Ç –ø—Ä–æ–º–∞—Ö.

–í –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –º—ã —Å—Ç—Ä–æ–∏–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ, —á—Ç–æ–±—ã –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –æ—Å–Ω–æ–≤–µ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –ù–∞–ø—Ä–∏–º–µ—Ä, –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç—å —Ü–µ–Ω—É –¥–æ–º–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ –µ–≥–æ –ø–ª–æ—â–∞–¥–∏, –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∫–æ–º–Ω–∞—Ç –∏ —Ç.–¥. –ò–Ω–æ–≥–¥–∞ –º–æ–¥–µ–ª—å –º–æ–∂–µ—Ç –ø—Ä–∏—Å–≤–∞–∏–≤–∞—Ç—å —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —ç—Ç–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º, —á—Ç–æ–±—ã —Ç–æ—á–Ω–æ –ø–æ–¥—Å—Ç—Ä–æ–∏—Ç—å—Å—è –ø–æ–¥ –æ–±—É—á–∞—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ. –≠—Ç–æ –∏ –µ—Å—Ç—å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ.

–†–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è –¥–æ–±–∞–≤–ª—è–µ—Ç "—à—Ç—Ä–∞—Ñ" –∫ —ç—Ç–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º (–∏–ª–∏ –≤–µ—Å–∞–º), —á—Ç–æ–±—ã –º–æ–¥–µ–ª—å –Ω–µ –º–æ–≥–ª–∞ —Å–¥–µ–ª–∞—Ç—å –∏—Ö —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–∏–º–∏. –≠—Ç–æ –ø–æ—Ö–æ–∂–µ –Ω–∞ —Ç–æ, –∫–∞–∫ –µ—Å–ª–∏ –±—ã —Ç—Ä–µ–Ω–µ—Ä –¥–∞–≤–∞–ª –Ω–µ–±–æ–ª—å—à–æ–µ —É—Ç—è–∂–µ–ª–µ–Ω–∏–µ –Ω–∞ —Ä—É–∫—É, —á—Ç–æ–±—ã –±—Ä–æ—Å–∫–∏ –±—ã–ª–∏ –±–æ–ª–µ–µ —Ç–æ—á–Ω—ã–º–∏ –∏ —Å—Ç–∞–±–∏–ª—å–Ω—ã–º–∏.

–ï—Å–ª–∏ –≥–æ–≤–æ—Ä–∏—Ç—å –±–æ–ª–µ–µ —Ñ–æ—Ä–º–∞–ª—å–Ω–æ, —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è ‚Äî —ç—Ç–æ –º–µ—Ç–æ–¥–∏–∫–∞ –≤ –º–∞—à–∏–Ω–Ω–æ–º –æ–±—É—á–µ–Ω–∏–∏, –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –Ω–∞ —É–º–µ–Ω—å—à–µ–Ω–∏–µ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ –∑–∞ —Å—á—ë—Ç –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–≥–æ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –∏–ª–∏ —à—Ç—Ä–∞—Ñ–∞ –∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å. –û—Å–Ω–æ–≤–Ω–∞—è –∏–¥–µ—è —Å–æ—Å—Ç–æ–∏—Ç –≤ —Ç–æ–º, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –º–æ–¥–µ–ª—å –±–æ–ª–µ–µ –ø—Ä–æ—Å—Ç–æ–π –∏ –º–µ–Ω–µ–µ —Å–∫–ª–æ–Ω–Ω–æ–π –∫ –ø–æ–¥–≥–æ–Ω–∫–µ –ø–æ–¥ —Å–ª—É—á–∞–π–Ω—ã–µ —à—É–º—ã –≤ –¥–∞–Ω–Ω—ã—Ö.

## L1 (LASSO)

L1-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è —Ç–∞–∫–∂–µ –∏–∑–≤–µ—Å—Ç–Ω–∞ –∫–∞–∫ Lasso (Least Absolute Shrinkage and Selection Operator) —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è. –û–Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∞ –Ω–∞ –¥–æ–±–∞–≤–ª–µ–Ω–∏–∏ —à—Ç—Ä–∞—Ñ–∞, —Ä–∞–≤–Ω–æ–≥–æ —Å—É–º–º–µ –∞–±—Å–æ–ª—é—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –º–æ–¥–µ–ª–∏.



$$ \sum_{i=1}^n(y_i - \hat{y_i})^2 + \lambda ||w||_1  ‚Üí min$$

–≥–¥–µ:
$\sum_{i=1}^n(y_i - \hat{y_i})^2$  ‚Äî –≤—ã–±—Ä–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å, –∫–æ—Ç–æ—Ä—É—é –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ–º, $w = (w_1, ..., w_d)$ ‚Äî –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤.

–ù–æ –Ω–∏–∫—Ç–æ –∂–µ –Ω–µ –º–µ—à–∞–µ—Ç –≤–∞–º —Ä–∞—Å–ø–∏—Å–∞—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ $ \hat{y_i}$ –∫–∞–∫: $$ \hat{y_i} = w_0 + \sum_{j=1}^d {w_j x_{ij}}$$ –æ—Ç–∫—É–¥–∞ –∏ –ø–æ—è–≤–ª—è—é—Ç—Å—è –≤–µ—Å–∞. –¢–æ –µ—Å—Ç—å —Ä–∞—Å–ø–∏—Å–∞–ª–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –∫–∞–∫ —Å—É–º–º—É –≤–µ—Å–æ–≤, —É–º–Ω–æ–∂–µ–Ω–Ω—ã—Ö –Ω–∞ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

–¢–µ–ø–µ—Ä—å —Ä–∞—Å–ø–∏—à–µ–º —Å–∞–º —Ä–µ–≥—É–ª—è—Ä–∏–∑–∏—Ä—É—é—â–∏–π —á–ª–µ–Ω –∫–∞–∫ —Å—É–º–º—É –∞–±—Å–æ–ª—é—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ (—Å–æ–≥–ª–∞—Å–Ω–æ —Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–µ):

$$Œª||w||_1 = \lambda\sum_{j = 1}^{d} |w_j|$$  - L1-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è, —à—Ç—Ä–∞—Ñ

$Œª$ - –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª—è–µ—Ç –∫–æ–Ω—Ç—Ä–æ–ª–∏—Ä–æ–≤–∞—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ —à—Ç—Ä–∞—Ñ—É–µ–º –≤–µ—Å–∞. –ß–µ–º –±–æ–ª—å—à–µ –ª—è–º–±–¥–∞, —Ç–µ–º —Å–∏–ª—å–Ω–µ–µ –º–æ–¥–µ–ª—å —à—Ç—Ä–∞—Ñ—É–µ—Ç—Å—è –∑–∞ –≤–µ–ª–∏—á–∏–Ω—É –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –∏ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ. –ï—Å–ª–∏ –∑–∞–Ω—É–ª–∏—Ç—å –ª—è–º–±–¥—É, –º—ã –ø–æ–ª—É—á–∏–º —Å–∞–º—É—é –æ–±—ã—á–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –ø–æ—Ç–µ—Ä—å –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ ‚Äî —Å–∞–º—É—é –æ–±—ã—á–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é.

> –û–±–∞ —Å–ø–æ—Å–æ–±–∞ —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏ –ø—Ä–∏–≤–æ–¥—è—Ç –∫ —É–º–µ–Ω—å—à–µ–Ω–∏—é –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏, –∑–∞—Å—Ç–∞–≤–ª—è—è –º–æ–¥–µ–ª—å –ø—ã—Ç–∞—Ç—å—Å—è —É–º–µ–Ω—å—à–∏—Ç—å —ç—Ç—É —Å—É–º–º—É.
–û–¥–Ω–∞–∫–æ —ç—Ç–æ—Ç —à—Ç—Ä–∞—Ñ –ø–æ–º–æ–≥–∞–µ—Ç –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—Ç–∏—Ç—å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ –∏ –º–æ–∂–µ—Ç –¥–∞–∂–µ –ø–æ–ª–Ω–æ—Å—Ç—å—é –∏—Å–∫–ª—é—á–∏—Ç—å –Ω–µ—Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏, –¥–µ–ª–∞—è –≤–µ—Å–∞ —Ä–∞–≤–Ω—ã–º–∏ –Ω—É–ª—é, –∏ **—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è –õ–∞—Å—Å–æ –∏–º–µ–µ—Ç —Ç–µ–Ω–¥–µ–Ω—Ü–∏—é –∑–∞–Ω—É–ª—è—Ç—å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Å–∏–ª—å–Ω–µ–µ**, —á–µ–º L2-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è.

## L2 (Ridge)

–ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ–µ –º–æ–∂–Ω–æ —Ä–∞—Å–ø–∏—Å–∞—Ç—å –∏ –¥–ª—è —Ä–∏–¥–∂-—Ä–µ–≥—Ä–µ—Å—Å–∏–∏. L2-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è –æ—Å–Ω–æ–≤–∞–Ω–∞ –Ω–∞ –¥–æ–±–∞–≤–ª–µ–Ω–∏–∏ —à—Ç—Ä–∞—Ñ–∞, —Ä–∞–≤–Ω–æ–≥–æ —Å—É–º–º–µ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –∑–Ω–∞—á–µ–Ω–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –º–æ–¥–µ–ª–∏.


$$\sum_{i=1}^n(y_i - \hat{y_i})^2 + \lambda ||w||_2^2 ‚Üí min $$

–≥–¥–µ: $\sum_{i=1}^n(y_i - \hat{y_i})^2$  - –≤–∞—à–∞ —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å, –∫–æ—Ç–æ—Ä—É—é –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ–º,
$x = (x_1,... , x_d)$ ‚Äî –≤–µ–∫—Ç–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, $w = (w_1, ..., w_d)$ ‚Äî –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤.

$$Œª||w||_2^2 = \lambda\sum_{j = 1}^{d} w_j^2$$  - L2-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è, —à—Ç—Ä–∞—Ñ

$Œª$ - –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª—è–µ—Ç –∫–æ–Ω—Ç—Ä–æ–ª–∏—Ä–æ–≤–∞—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ —à—Ç—Ä–∞—Ñ—É–µ–º –≤–µ—Å–∞.

–û–¥–Ω–∞–∫–æ —Ä–µ–≥—Ä–µ—Å—Å–∏—è –†–∏–¥–∂ –∏–º–µ–µ—Ç —Ç–µ–Ω–¥–µ–Ω—Ü–∏—é –∑–∞–Ω–∏–∂–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤, –Ω–æ –Ω–µ –ø–æ–ª–Ω–æ—Å—Ç—å—é —Å–≤–æ–¥–∏—Ç—å –∏—Ö –∫ –Ω—É–ª—é.


## –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è

–ö–æ–≥–¥–∞ –º—ã —Ä–µ—à–∞–µ–º –∑–∞–¥–∞—á—É –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –≤ –º–∞—à–∏–Ω–Ω–æ–º –æ–±—É—á–µ–Ω–∏–∏, –æ—Å–Ω–æ–≤–Ω–∞—è —Ü–µ–ª—å ‚Äî –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –ø–æ—Ç–µ—Ä—å. –ß–µ–º –ª—É—á—à–µ –º—ã —ç—Ç–æ —Å–¥–µ–ª–∞–µ–º, —Ç–µ–º –±–ª–∏–∂–µ –º—ã –∫ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–º—É —Ä–µ—à–µ–Ω–∏—é, —Ç–æ –µ—Å—Ç—å –∫ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é —Ñ—É–Ω–∫—Ü–∏–∏. –¢–µ–ø–µ—Ä—å –¥–∞–≤–∞–π—Ç–µ –ø–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è –≤–ª–∏—è–µ—Ç –Ω–∞ —ç—Ç–æ—Ç –ø—Ä–æ—Ü–µ—Å—Å —Å –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–π —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è.

–í –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å ‚Äî —ç—Ç–æ RSS (—Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π), –∫–æ—Ç–æ—Ä—É—é –Ω—É–∂–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å. –ß—Ç–æ–±—ã –ø—Ä–æ—â–µ –ø–æ–Ω—è—Ç—å —Å–∏—Ç—É–∞—Ü–∏—é, –ø—Ä–µ–¥—Å—Ç–∞–≤–∏–º, —á—Ç–æ —É –Ω–∞—Å –µ—Å—Ç—å –≤—Å–µ–≥–æ –æ–¥–∏–Ω –ø—Ä–∏–∑–Ω–∞–∫ –≤ –¥–∞–Ω–Ω—ã—Ö. $x_0$ ‚Äî —ç—Ç–æ —Ñ–∏–∫—Ç–∏–≤–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫, –∫–æ—Ç–æ—Ä—ã–π —Å–æ–∑–¥–∞–ª–∏ –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ.

–ù–∞ –≥—Ä–∞—Ñ–∏–∫–µ –ø–æ–∫–∞–∑–∞–Ω–æ —Ä–µ—à–µ–Ω–∏–µ $\hat{w}$, –∫–æ—Ç–æ—Ä–æ–µ —è–≤–ª—è–µ—Ç—Å—è —ç—Ç–∞–ª–æ–Ω–Ω—ã–º, –Ω–æ –ø–æ–ª—É—á–µ–Ω–æ –±–µ–∑ —É—á–µ—Ç–∞ —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏. –≠—Ç–æ —Ç–∞–∫ –Ω–∞–∑—ã–≤–∞–µ–º—ã–π "–∏–¥–µ–∞–ª—å–Ω—ã–π" –º–∏–Ω–∏–º—É–º –¥–ª—è –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å.

–§—É–Ω–∫—Ü–∏—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö (–¥–≤—É—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤) –æ–±—Ä–∞–∑—É–µ—Ç –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å –≤ —Ç—Ä–µ—Ö–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ. –ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–µ $(w_1, w_2)$ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –≤—ã—Å–æ—Ç—É —ç—Ç–æ–π –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ ‚Äî –∑–Ω–∞—á–µ–Ω–∏–µ –æ—à–∏–±–∫–∏. –ï—Å–ª–∏ –º—ã –ø–æ—Å–º–æ—Ç—Ä–∏–º —Å–≤–µ—Ä—Ö—É –Ω–∞ —ç—Ç—É –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å, —Ç–æ —É–≤–∏–¥–∏–º –∫–æ–Ω—Ç—É—Ä–Ω—ã–µ –ª–∏–Ω–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º —Ñ—É–Ω–∫—Ü–∏–∏. –ö—Ä–∞—Å–Ω—ã–µ —ç–ª–ª–∏–ø—Å—ã –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –∏–ª–ª—é—Å—Ç—Ä–∏—Ä—É—é—Ç –≤–∏–¥ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å —Å–≤–µ—Ä—Ö—É ‚Äî –∫–∞–∫ –µ—Å–ª–∏ –±—ã –º—ã —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–ª–∏ —Å—Ä–µ–∑—ã —É—Ä–æ–≤–Ω–µ–π. –≠—Ç–æ –ø–æ—Ö–æ–∂–µ –Ω–∞ —Ç–æ–ø–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–µ –∫–∞—Ä—Ç—ã, –≥–¥–µ –ª–∏–Ω–∏–∏ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç –≤—ã—Å–æ—Ç—É, —Ç–æ–ª—å–∫–æ –∑–¥–µ—Å—å –æ–Ω–∏ —É–∫–∞–∑—ã–≤–∞—é—Ç –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –∫ –º–∏–Ω–∏–º—É–º—É.

–¢–µ–ø–µ—Ä—å –æ —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏. –†–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è –≤–≤–æ–¥–∏—Ç —à—Ç—Ä–∞—Ñ—ã –∏–ª–∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –¥–ª—è –Ω–∞—à–µ–π –º–æ–¥–µ–ª–∏. –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏ —ç—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≤—Å–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã—Ö–æ–¥—è—Ç –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –∑–∞–¥–∞–Ω–Ω–æ–π —Ñ–æ—Ä–º—ã (–∫—Ä—É–≥–∞ –∏–ª–∏ –∫–≤–∞–¥—Ä–∞—Ç–∞), –Ω–∞–∫–∞–∑—ã–≤–∞—é—Ç—Å—è. –≠—Ç–∏ —Ñ–æ—Ä–º—ã –æ–ø—Ä–µ–¥–µ–ª—è—é—Ç—Å—è —Ç–∏–ø–æ–º —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏ ‚Äî L1 (Lasso) –∏–ª–∏ L2 (Ridge) ‚Äî –∏ –∑–∞–¥–∞—é—Ç –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –Ω–∞ —Å—É–º–º—É –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤. –ï—Å–ª–∏ —ç—Ç–∞ —Å—É–º–º–∞ –ø—Ä–µ–≤—ã—à–∞–µ—Ç —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—ã–π –ø–æ—Ä–æ–≥, —Å—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç $Œª$, –∫–æ—Ç–æ—Ä—ã–π —Ä–µ–∑–∫–æ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å, –¥–µ–ª–∞—è —Ä–µ—à–µ–Ω–∏–µ –Ω–µ–æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–º.

–ß—Ç–æ–±—ã —Ä–µ—à–µ–Ω–∏–µ —Å—á–∏—Ç–∞–ª–æ—Å—å –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–º —Å —É—á–µ—Ç–æ–º —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏, –æ–Ω–æ –¥–æ–ª–∂–Ω–æ –Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –≤–Ω—É—Ç—Ä–∏ –æ–±–ª–∞—Å—Ç–∏, –∑–∞–¥–∞–≤–∞–µ–º–æ–π —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–µ–π. –í –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ, –∫–∞–∫ —Ç–æ–ª—å–∫–æ —Ä–µ—à–µ–Ω–∏–µ –≤—ã—Ö–æ–¥–∏—Ç –∑–∞ –ø—Ä–µ–¥–µ–ª—ã —ç—Ç–æ–π –æ–±–ª–∞—Å—Ç–∏, –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç $Œª$ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é –ø–æ—Ç–µ—Ä—å. –≠—Ç–æ –∑–∞—Å—Ç–∞–≤–ª—è–µ—Ç –∞–ª–≥–æ—Ä–∏—Ç–º –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –∫–æ–º–ø—Ä–æ–º–∏—Å—Å–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ, –º–∏–Ω–∏–º–∏–∑–∏—Ä—É—é—â–µ–µ –ø–æ—Ç–µ—Ä–∏ –∏ –æ—Å—Ç–∞—é—â–µ–µ—Å—è –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–æ–Ω–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏.

## ElasticNet

ElasticNet - —ç—Ç–æ –∫–æ–º–±–∏–Ω–∞—Ü–∏—è –∏–∑ –†–∏–¥–∂- –∏ –õ–∞—Å—Å–æ-—Ä–µ–≥—Ä–µ—Å—Å–∏–∏.

![](https://alexanderdyakonov.files.wordpress.com/2019/10/f8.png?w=700)

–í –±–æ–ª–µ–µ –ø—Ä–∏–≤—ã—á–Ω–æ–π —Ñ–æ—Ä–º–µ –µ–≥–æ –º–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å —Ç–∞–∫:

$$(\lambda_1 \sum|w_j| + \lambda_2 \sum w_j^2)$$

–£ —ç—Ç–æ–π —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏ –µ—Å—Ç—å –Ω–∞—Å—Ç—Ä–æ–∏–≤–∞–µ–º—ã–π –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä ‚Äî l1_ratio, —É–∫–∞–∑—ã–≤–∞—é—â–∏–π –Ω–∞ –¥–æ–ª—é L1 –∏ L2 —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏. l1_ratio = 0, —Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ L2 —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è, –ø—Ä–∏ l1_ratio = 1 - —Ç–æ–ª—å–∫–æ L1. –ü—Ä–∏ 0 < l1_ratio < 1 - –∫–æ–º–±–∏–Ω–∞—Ü–∏—è L1 –∏ L2.

# –ü–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

**–ü–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è** ‚Äî —Ä–∞—Å—à–∏—Ä–µ–Ω–∏–µ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –ø–æ–∑–≤–æ–ª—è—é—â–µ–µ –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞—Ç—å –Ω–µ–ª–∏–Ω–µ–π–Ω—ã–µ —Å–≤—è–∑–∏ –∑–∞ —Å—á–µ—Ç –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–∫–≤–∞–¥—Ä–∞—Ç, –∫—É–± –∏ —Ç.–¥.):

$y_{pred}(x)=w_0+\sum_{j = 1}^{d} w_j x_j^k, k = \overline{1, n}$

$n$ - —Å—Ç–µ–ø–µ–Ω—å –ø–æ–ª–∏–Ω–æ–º–∞

$w$ - –≤–µ—Å–∞/–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã

–ü—Ä–æ—Å—Ç–∞—è –ª–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π –Ω–µ —á—Ç–æ –∏–Ω–æ–µ, –∫–∞–∫ –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é –ø–µ—Ä–≤–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞. –í –æ—Ç–ª–∏—á–∏–µ –æ—Ç –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –≥–¥–µ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –∏ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π —è–≤–ª—è–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω–æ–π, –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –º–æ–∂–µ—Ç –æ–ø–∏—Å—ã–≤–∞—Ç—å –∫—Ä–∏–≤—ã–µ, –±–ª–∞–≥–æ–¥–∞—Ä—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—é –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–µ–ø–µ–Ω–µ–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

–ú–æ–∂–µ—Ç –≤–æ–∑–Ω–∏–∫–Ω—É—Ç—å –≤–æ–ø—Ä–æ—Å, –∞ –∫–∞–∫ —Ç–æ–≥–¥–∞ –∏—Å–∫–∞—Ç—å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏ –Ω–æ–≤—ã—Ö –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö? –ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ, —Ä–µ—à–µ–Ω–∏–µ –Ω–µ –±—É–¥–µ—Ç —Å–∏–ª—å–Ω–æ –æ—Ç–ª–∏—á–∞—Ç—å—Å—è. –ù–æ–≤—ã–µ –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å–æ–∑–¥–∞–Ω—ã –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ, –æ–¥–Ω–∞–∫–æ –æ–± —ç—Ç–æ–π "–∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ—Å—Ç–∏" –∑–Ω–∞—é—Ç —Ç–æ–ª—å–∫–æ —Å–æ–∑–¥–∞—Ç–µ–ª–∏. –°–∞–º–∞ –º–æ–¥–µ–ª—å —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –æ –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ—Å—Ç–∏ –Ω–∏—á–µ–≥–æ –Ω–µ –∑–Ω–∞–µ—Ç: —Ç–∞–∫, –¥–ª—è –Ω–µ—ë, –Ω–∞–ø—Ä–∏–º–µ—Ä, —Å–æ–∑–¥–∞–Ω–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫ $x_1^2$ —Ä–∞–≤–Ω–æ–∑–Ω–∞—á–µ–Ω –Ω–æ–≤–æ–º—É –≤–≤–æ–¥–∏–º–æ–º—É $x_{d+1}$, –∞ –æ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ—Å—Ç–∏ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –Ω–∏—á–µ–≥–æ –Ω–µ –∑–Ω–∞–µ—Ç. –ï—Å–ª–∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ñ–æ—Ä–º—É–ª—É, —Ç–æ –ø—Ä–∏ –≤–µ—Å–∞—Ö –Ω–µ—Ç —Å—Ç–µ–ø–µ–Ω–∏ –ø–æ–ª–∏–Ω–æ–º–∞ - —Å—Ç–µ–ø–µ–Ω—å—é –Ω–∞–¥–µ–ª—è—é—Ç—Å—è –∏—Å–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–∏–∑–Ω–∞–∫–∏ $x$.

**–° —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º —Å—Ç–µ–ø–µ–Ω–∏ –ø–æ–ª–∏–Ω–æ–º–∞ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –∑–∞–¥–∞—á–∏ –±—ã—Å—Ç—Ä–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç**, —á—Ç–æ –º–æ–∂–µ—Ç –ø—Ä–∏–≤–æ–¥–∏—Ç—å –∫ "–ø—Ä–æ–∫–ª—è—Ç–∏—é —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏". –≠—Ç–æ –æ—Å–æ–±–µ–Ω–Ω–æ –∞–∫—Ç—É–∞–ª—å–Ω–æ –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ –º–æ–¥–µ–ª–µ–π —Å –º–Ω–æ–≥–∏–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏, —Ç–∞–∫ –∫–∞–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã—Ö –∫–æ–º–±–∏–Ω–∞—Ü–∏–π –±—ã—Å—Ç—Ä–æ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è.

–û–¥–Ω–∞ –∏–∑ –ø—Ä–æ–±–ª–µ–º –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ ‚Äî —Å–∫–ª–æ–Ω–Ω–æ—Å—Ç—å –∫ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—é –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ –ø–æ–ª–∏–Ω–æ–º–æ–≤ –≤—ã—Å–æ–∫–æ–π —Å—Ç–µ–ø–µ–Ω–∏. –ú–æ–¥–µ–ª—å –º–æ–∂–µ—Ç –∏–¥–µ–∞–ª—å–Ω–æ –ø–æ–¥–æ–≥–Ω–∞—Ç—å—Å—è –ø–æ–¥ –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É, –Ω–æ –ø–ª–æ—Ö–æ –æ–±–æ–±—â–∞—Ç—å—Å—è –Ω–∞ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –≠—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –∏–∑-–∑–∞ —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–æ–π –≥–∏–±–∫–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—á–∏–Ω–∞–µ—Ç "–∑–∞–ø–æ–º–∏–Ω–∞—Ç—å" —à—É–º –¥–∞–Ω–Ω—ã—Ö –≤–º–µ—Å—Ç–æ —Ç–æ–≥–æ, —á—Ç–æ–±—ã –≤—ã—è–≤–ª—è—Ç—å –æ–±—â–∏–µ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏. –ß—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å —ç—Ç–æ–≥–æ, –≤–∞–∂–Ω–æ –∫–æ–Ω—Ç—Ä–æ–ª–∏—Ä–æ–≤–∞—Ç—å —Å—Ç–µ–ø–µ–Ω—å –ø–æ–ª–∏–Ω–æ–º–∞ –∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—é.

–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –ø–æ–ª–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å–ª–æ–∂–Ω–µ–µ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π. –ö–∞–∂–¥–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ —Å—Ç–µ–ø–µ–Ω–∏ –ø–æ–ª–∏–Ω–æ–º–∞ –º–µ–Ω—è–µ—Ç —Ñ–æ—Ä–º—É –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏, –∏ —ç—Ç–æ –¥–µ–ª–∞–µ—Ç –∞–Ω–∞–ª–∏–∑ –≤–ª–∏—è–Ω–∏—è –æ—Ç–¥–µ–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –º–µ–Ω–µ–µ –æ—á–µ–≤–∏–¥–Ω—ã–º.

# –ú–µ—Ç—Ä–∏–∫–∏

–ú–µ—Ç—Ä–∏–∫–∏ VS —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å:

* —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å –æ–ø—Ç–∏–º–∏–∑–∏—Ä—É–µ—Ç—Å—è/–º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ—Ç—Å—è –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ –æ–±—É—á–µ–Ω–∏—è

* –º–µ—Ç—Ä–∏–∫–∏ —Å–ª—É–∂–∞—Ç –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞/–ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏

–ú–µ—Ç—Ä–∏–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ ‚Äî —ç—Ç–æ –æ—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –Ω–∞—à–µ–π –º–æ–¥–µ–ª–∏. –ü—Ä–∏—á–µ–º –¥–ª—è –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π –∑–∞–¥–∞—á–∏ –º–æ–∂–µ—Ç –≤–æ–∑–Ω–∏–∫–∞—Ç—å —Ü–µ–ª–∞—è –∏–µ—Ä–∞—Ä—Ö–∏—è –º–µ—Ç—Ä–∏–∫. –≠—Ç–æ –≤–Ω–µ—à–Ω–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π –∫–∞—á–µ—Å—Ç–≤–∞, –∑–∞–≤–∏—Å—è—â–∏–π –æ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π, –Ω–æ –Ω–µ –æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏.

–§—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å ‚Äî –∫—Ä–∏—Ç–µ—Ä–∏–π ¬´–æ–±—É—á–∞–µ–º–æ—Å—Ç–∏¬ª –Ω–∞—à–µ–π –º–æ–¥–µ–ª–∏. –í–æ–∑–Ω–∏–∫–∞–µ—Ç, –∫–æ–≥–¥–∞ –º—ã –ø–µ—Ä–µ—Ö–æ–¥–∏–º –æ—Ç –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ –∫ –∑–∞–¥–∞—á–µ –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏, —Ç–æ –µ—Å—Ç—å –ø–æ–∏—Å–∫—É –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏.



 $$ MSE = \frac{1}{n} \sum_{i=1}^n(y_i - \hat{y_i})^2 $$

–ï—ë –º—ã —É–∂–µ –∑–∞—Ç—Ä–∞–≥–∏–≤–∞–ª–∏ - —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ (Mean Squared Error), –∏–Ω–∞—á–µ - —Å—Ä–µ–¥–Ω—è—è —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –æ—Ç —ç—Ç–∞–ª–æ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞. –ß–µ–º –±–ª–∏–∂–µ –æ–Ω–∞ –∫ –Ω—É–ª—é, —Ç–µ–º –ª—É—á—à–µ –º–æ–¥–µ–ª—å –∞–¥–∞–ø—Ç–∏—Ä–æ–≤–∞–ª–∞—Å—å –ø–æ–¥ –¥–∞–Ω–Ω—ã–µ.

MSE –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –¥—Ä—É–≥–∏—Ö –º–µ—Ç—Ä–∏–∫ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å–≤–æ–µ–π —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –∫ –±–æ–ª—å—à–∏–º –æ—à–∏–±–∫–∞–º –±–ª–∞–≥–æ–¥–∞—Ä—è –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–º—É —à—Ç—Ä–∞—Ñ—É, –∫–æ—Ç–æ—Ä—ã–π –æ–Ω –Ω–∞–∫–ª–∞–¥—ã–≤–∞–µ—Ç. –í –æ—Ç–ª–∏—á–∏–µ –æ—Ç –º–µ—Ç—Ä–∏–∫, —Ç–∞–∫–∏—Ö –∫–∞–∫ MAE (Mean Absolute Error), –∫–æ—Ç–æ—Ä–∞—è –∏–∑–º–µ—Ä—è–µ—Ç —Å—Ä–µ–¥–Ω—é—é –∞–±—Å–æ–ª—é—Ç–Ω—É—é –æ—à–∏–±–∫—É, MSE –≤–æ–∑–≤–æ–¥–∏—Ç –∫–∞–∂–¥—É—é –æ—à–∏–±–∫—É –≤ –∫–≤–∞–¥—Ä–∞—Ç, —á—Ç–æ –¥–µ–ª–∞–µ—Ç –µ–≥–æ –±–æ–ª–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–º –∫ –∫—Ä—É–ø–Ω—ã–º –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è–º.

–≠—Ç–∞ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –ø—Ä–∏—Ä–æ–¥–∞ MSE –ø—Ä–∏–≤–æ–¥–∏—Ç –∫ —Ç–æ–º—É, —á—Ç–æ –±–æ–ª—å—à–∏–µ –æ—à–∏–±–∫–∏ –æ–∫–∞–∑—ã–≤–∞—é—Ç –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å—à–µ–µ –≤–ª–∏—è–Ω–∏–µ –Ω–∞ –∏—Ç–æ–≥–æ–≤—É—é –æ—Ü–µ–Ω–∫—É, —á–µ–º –º–∞–ª–µ–Ω—å–∫–∏–µ –æ—à–∏–±–∫–∏. –≠—Ç–æ —Å–≤–æ–π—Å—Ç–≤–æ –ø–æ–ª–µ–∑–Ω–æ, –∫–æ–≥–¥–∞ –≤–∞–∂–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∫—Ä—É–ø–Ω—ã–µ –æ—à–∏–±–∫–∏, —Ç–∞–∫ –∫–∞–∫ –º–æ–¥–µ–ª—å, –æ–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –ø–æ MSE, –±—É–¥–µ—Ç —Å—Ç—Ä–µ–º–∏—Ç—å—Å—è –∏–∑–±–µ–≥–∞—Ç—å –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π.

–û–¥–Ω–∞–∫–æ —ç—Ç–∞ –∂–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç—å –¥–µ–ª–∞–µ—Ç MSE –º–µ–Ω–µ–µ –ø–æ–¥—Ö–æ–¥—è—â–∏–º –≤ —Å–∏—Ç—É–∞—Ü–∏—è—Ö, –≥–¥–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤—ã–±—Ä–æ—Å—ã. –í—ã–±—Ä–æ—Å—ã –º–æ–≥—É—Ç —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ —É–≤–µ–ª–∏—á–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ MSE, —á—Ç–æ —É—Å–ª–æ–∂–Ω—è–µ—Ç –æ–±—ä–µ–∫—Ç–∏–≤–Ω–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π –∏ –º–æ–∂–µ—Ç –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ —Ç–æ–º—É, —á—Ç–æ –º–æ–¥–µ–ª—å –±—É–¥–µ—Ç —á—Ä–µ–∑–º–µ—Ä–Ω–æ –Ω–∞—Å—Ç—Ä–æ–µ–Ω–∞ –Ω–∞ –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏—é –≤–ª–∏—è–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤, –∏–≥–Ω–æ—Ä–∏—Ä—É—è –ø—Ä–∏ —ç—Ç–æ–º –æ—Å–Ω–æ–≤–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö.

 $$ RMSE = \sqrt{MSE} $$

RMSE ‚Äî –∫–æ—Ä–µ–Ω—å –∏–∑ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π –æ—à–∏–±–∫–∏. –û—Å–Ω–æ–≤–Ω–æ–µ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ RMSE –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ –æ–Ω–∞ –∏–∑–º–µ—Ä—è–µ—Ç –æ—à–∏–±–∫—É –≤ —Ç–µ—Ö –∂–µ –µ–¥–∏–Ω–∏—Ü–∞—Ö, —á—Ç–æ –∏ –∑–∞–≤–∏—Å–∏–º–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è, —á—Ç–æ –¥–µ–ª–∞–µ—Ç –µ—ë –ª–µ–≥–∫–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä—É–µ–º–æ–π. –ë–æ–ª–µ–µ —Ç–æ–≥–æ, RMSE –¥–∞–µ—Ç –±–æ–ª—å—à–∏–π –≤–µ—Å –±–æ–ª—å—à–∏–º –æ—à–∏–±–∫–∞–º –∏–∑-–∑–∞ –æ–ø–µ—Ä–∞—Ü–∏–∏ –≤–æ–∑–≤–µ–¥–µ–Ω–∏—è –≤ –∫–≤–∞–¥—Ä–∞—Ç –∏ –ø–æ—Å–ª–µ–¥—É—é—â–µ–≥–æ –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –∫–æ—Ä–Ω—è. –û–¥–Ω–∞–∫–æ –∫–∞–∫ –∏ –ª—é–±–∞—è –º–µ—Ç—Ä–∏–∫–∞, RMSE –Ω–µ –ª–∏—à–µ–Ω–∞ –Ω–µ–¥–æ—Å—Ç–∞—Ç–∫–æ–≤. –ù–∞–ø—Ä–∏–º–µ—Ä, –æ–Ω–∞ —Ç–∞–∫–∂–µ –º–æ–∂–µ—Ç –±—ã—Ç—å —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞ –∫ –≤—ã–±—Ä–æ—Å–∞–º, –ø–æ—Å–∫–æ–ª—å–∫—É –±–æ–ª—å—à–∏–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –º–æ–≥—É—Ç –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ —É–≤–µ–ª–∏—á–∏–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ RMSE –∏ –∏—Å–∫–∞–∂–∞—Ç—å –æ–±—â—É—é –æ—Ü–µ–Ω–∫—É –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏.


 $$ MAE = \frac{1}{n} \sum_{i=1}^n|y_i - \hat{y_i}| $$

‚Äî —É—Å—Ä–µ–¥–Ω–µ–Ω–Ω–∞—è —Å—É–º–º–∞ —Ä–∞–∑–Ω–∏—Ü—ã –º–µ–∂–¥—É —ç—Ç–∞–ª–æ–Ω–Ω—ã–º –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º, –≤–∑—è—Ç–æ–π –ø–æ –º–æ–¥—É–ª—é.

MAE (Mean Absolute Error - —Å—Ä–µ–¥–Ω—è—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –æ—à–∏–±–∫–∞) –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è —Ç–µ–º, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –æ—à–∏–±–æ–∫, –∞ –Ω–µ –∏—Ö –∫–≤–∞–¥—Ä–∞—Ç—ã. –≠—Ç–æ –¥–µ–ª–∞–µ—Ç MAE –º–µ–Ω–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–º –∫ –≤—ã–±—Ä–æ—Å–∞–º –∏ –∫—Ä—É–ø–Ω—ã–º –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è–º, –ø–æ—Å–∫–æ–ª—å–∫—É –∫–∞–∂–¥–∞—è –æ—à–∏–±–∫–∞ –≤–Ω–æ—Å–∏—Ç –ª–∏–Ω–µ–π–Ω—ã–π –≤–∫–ª–∞–¥ –≤ –æ–±—â—É—é –æ—Ü–µ–Ω–∫—É, –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –µ—ë –≤–µ–ª–∏—á–∏–Ω—ã.

–û—Å–Ω–æ–≤–Ω–∞—è –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç—å MAE –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ –æ–Ω–∞ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —à—Ç—Ä–∞—Ñ—É–µ—Ç –≤—Å–µ –æ—à–∏–±–∫–∏, –±–µ–∑ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–≥–æ –≤–µ—Å–∞ –¥–ª—è –±–æ–ª—å—à–∏—Ö –æ—à–∏–±–æ–∫, –∫–∞–∫ —ç—Ç–æ –¥–µ–ª–∞–µ—Ç MSE. –≠—Ç–æ –¥–µ–ª–∞–µ—Ç MAE –±–æ–ª–µ–µ —É—Å—Ç–æ–π—á–∏–≤–æ–π –∫ –≤—ã–±—Ä–æ—Å–∞–º, –ø–æ—Å–∫–æ–ª—å–∫—É –∫—Ä—É–ø–Ω—ã–µ –æ—à–∏–±–∫–∏ –Ω–µ –æ–∫–∞–∑—ã–≤–∞—é—Ç —á—Ä–µ–∑–º–µ—Ä–Ω–æ–≥–æ –≤–ª–∏—è–Ω–∏—è –Ω–∞ –∏—Ç–æ–≥–æ–≤—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç. –ë–ª–∞–≥–æ–¥–∞—Ä—è —ç—Ç–æ–º—É —Å–≤–æ–π—Å—Ç–≤—É MAE –ª—É—á—à–µ –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è —Å–∏—Ç—É–∞—Ü–∏–π, –≥–¥–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤—ã–±—Ä–æ—Å—ã, –∏ –≤–∞–∂–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å —Å—Ä–µ–¥–Ω—é—é –æ—à–∏–±–∫—É –±–µ–∑ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏—Å–∫–∞–∂–µ–Ω–∏—è –∏–∑-–∑–∞ —Ä–µ–¥–∫–∏—Ö, –Ω–æ –±–æ–ª—å—à–∏—Ö –æ—à–∏–±–æ–∫.

$$ MAPE = \frac{1}{n} \sum_{i=1}^n\left|{\frac{y_i - \hat{y_i}}{y_i}}\right| $$

–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è, —á—Ç–æ–±—ã —É—á–∏—Ç—ã–≤–∞—Ç—å –º–∞—Å—à—Ç–∞–± —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π. –¢–æ –µ—Å—Ç—å –º—ã –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç,–Ω–∞ –∫–æ—Ç–æ—Ä—ã–π –Ω–∞—à–∞ –º–æ–¥–µ–ª—å –æ—à–∏–±–∞–µ—Ç—Å—è.

MAPE ‚Äî –º–µ—Ç—Ä–∏–∫–∞ —Å –æ—á–µ–Ω—å –ø—Ä–æ—Å—Ç–æ–π –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏–µ–π, –∏–∑–º–µ—Ä—è–µ–º–∞—è –≤ –¥–æ–ª—è—Ö –∏–ª–∏ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö. –ï—Å–ª–∏ —É –≤–∞—Å –ø–æ–ª—É—á–∏–ª–æ—Å—å, –Ω–∞–ø—Ä–∏–º–µ—Ä, —á—Ç–æ MAPE = n%, —Ç–æ —ç—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ –æ—à–∏–±–∫–∞ —Å–æ—Å—Ç–∞–≤–∏–ª–∞ n% –æ—Ç —Ä–µ–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π. –û–¥–Ω–∞–∫–æ —Ç–∞–∫–∞—è –º–µ—Ç—Ä–∏–∫–∞ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω–∞: –ª—é–±–æ–µ —Å–∏–ª—å–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –º–æ–∂–µ—Ç –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –ø–æ–≤–ª–∏—è—Ç—å –Ω–∞ —Ä–∞–∑–º–µ—Ä –æ—à–∏–±–∫–∏.

$$ R^2 = 1 - \frac{\sum_{i=1}^n(y_i - \hat{y_i})^2}{\sum_{i=1}^n(y_i - \overline{y})^2} $$

$R^2 - $ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–∏–Ω–∞—Ü–∏–∏, –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫—É—é —á–∞—Å—Ç—å –∏–∑–º–µ–Ω—á–∏–≤–æ—Å—Ç–∏ –Ω–∞–±–ª—é–¥–∞–µ–º–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –º–æ–∂–Ω–æ –æ–±—ä—è—Å–Ω–∏—Ç—å —Å –ø–æ–º–æ—â—å—é –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏. –í –æ—Ç–ª–∏—á–∏–µ –æ—Ç –¥—Ä—É–≥–∏—Ö –º–µ—Ç—Ä–∏–∫, —ç—Ç—É –º—ã —Å—Ç—Ä–µ–º–∏–º—Å—è –º–∞–∫—Å–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ –º–æ–¥–µ–ª–∏, —Ç. –∫. –æ–Ω–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞ –ø—Ä–∏ MSE = 0, –Ω–∞–ø—Ä–∏–º–µ—Ä. –ó–Ω–∞—á–µ–Ω–∏–µ R2 –º–µ–Ω—å—à–µ 0 –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º–∞—è –º–æ–¥–µ–ª—å —Ö—É–∂–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–æ–π. –£ –∏–¥–µ–∞–ª—å–Ω–æ–≥–æ —Ä–µ—à–∞—é—â–µ–≥–æ –ø—Ä–∞–≤–∏–ª–∞ $ R^2 = 1$.

–õ–µ–≥–∫–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å ‚Äî –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –¥–æ–ª—é –∏–∑–º–µ–Ω–µ–Ω–∏–π, –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω—ã—Ö –≤–ª–∏—è–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–∞ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é. –û–¥–Ω–∞–∫–æ –Ω–µ —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è –ø—Ä–∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–∏ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –∏—Ö —Å–≤—è–∑–∏ —Å —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.
