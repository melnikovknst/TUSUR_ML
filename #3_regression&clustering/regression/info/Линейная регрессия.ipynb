# Что есть задача регрессии

Задача регрессии в машинном обучении относится к классу задач, где целью является прогнозирование **непрерывной числовой переменной**. Другими словами, регрессия направлена на построение модели, способной предсказывать значения непрерывной зависимой переменной на основе входных данных. Это отличается от задачи классификации, где целью является прогнозирование категориальной переменной.

# Линейная регрессия

## Линейная регрессия и многомерная линейная регрессия

Линейная регрессия, пожалуй, один из самых известных и наиболее понятных алгоритмов в статистике и машинном обучении. Линейная регрессия — это линейная модель, которая предполагает линейную связь между входными переменными ( x ) и единственной выходной переменной ( y ). Цель линейной регрессии — научить линейную модель прогнозировать новый y с учетом ранее невиданного x с как можно меньшей ошибкой.

Например, на картинке мы наблюдаем корреляцию - то, как располагаются синие точки, имеет определенное направление, и грубо это можно описать направлением прямой линейной регрессии.

common.png

Рассмотрим самый простой случай - когда есть один признак и целевое значение. Линейная регрессия предполагает, что между переменными и целевой существует линейная связь, то есть зависимость вида:

$$y=w_0+w_1 x_1$$

$y$ - эталонное целевое значение,

$x_1$ - значение признака,

$w_1, w_0$ - коэффициенты регрессии (они показывают влияние каждой независимой переменной на результат).

**Линейность** означает, что зависимость между независимыми переменными (𝑥) и зависимой переменной (𝑦) можно выразить как линейную комбинацию независимых переменных. Это подразумевает, что каждая независимая переменная влияет на зависимую переменную прямо пропорционально её коэффициенту регрессии.

На этом моменте можно вспомнить уравнение прямой, знакомое со школы:
$$y = kx + b$$




Точно также, как и в уравнении простой прямой, $w_0$ (intercept) $w_1$ (slope.)

**Intercept** — это точка, в которой прямая пересекает ось 𝑦 (вертикальную ось) при 𝑥=0.

* Визуально: это смещение линии вверх или вниз на графике.

* Интерпретация: если все независимые переменные равны нулю (𝑥=0), значение зависимой переменной 𝑦 будет равно $w_0$​ .

**Slope** (угол наклона) определяет крутизну наклона линии. Это коэффициент при
𝑥, который показывает, как быстро изменяется зависимая переменная 𝑦 при изменении независимой переменной 𝑥.

* Визуально: это изменение наклона линии — насколько она поднимается или опускается при изменении 𝑥.

* Интерпретация: slope ($w_1$) показывает, насколько изменяется 𝑦 при изменении
𝑥 на одну единицу.

> **Intercept** ($w_0$) влияет на то, где линия пересекает ось 𝑦, то есть смещает линию вверх или вниз.

> **Slope** ($w_1$) определяет угол наклона линии. Если $w_1>0$, линия будет наклонена вверх (положительный наклон), если $w_1<0$, линия будет наклонена вниз (отрицательный наклон).

image.png

**Множественная** линейная регрессия расширяет эту модель, добавляя несколько независимых переменных, и сводятся к суммированию значений признаков с некоторыми коэффициентами модели регрессии:

$y = w_1x_1 + w_2x_2 + .. + w_dx_d + w_0$, или

$$y = f(x)=w_0+\sum_{j = 1}^{d} w_j x_j$$

Параметрами модели многомерной линейной регрессии являются веса или коэффициенты $w_j$.
Вес $w_0$ также называется свободным коэффициентом или сдвигом (bias, intercept). $d$ - количество признаков в данных, а вектор $x_1, x_2... x_d$ - вектор значений признаков.

Раз мы вывели формулу зависимости, то через неё можем описать это выражение *для каждого* наблюдения:

$$ y_i = w_0 + \sum_{j=1}^d {w_j x_{ij}} $$, где $i = \overline{1, n}$. Мы добавили при каждом $y$ индекс строки $i$ - таким образом, при имеющихся идеальных весах регрессии, мы можем вывести значения для каждой целевой переменной $y$ значение на основе входных данных $x$ этой строки.








## Способ определения прямой

Однако прямую линейной регрессии можно провести по-разному, например, как на картинке для случая одного признака. Нам же надо каким-то образом сравнивать наши модели регрессии и находить наилучшую?

Очевидно, что необходимо найти какой-то математический подход к тому, чтобы наиболее оптимально провести прямую как в двумерном, так и в многомерном пространстве. Математически это обозначает наиболее оптимально подобрать коэффициенты модели линейной регрессии.

Для этого мы можем подбирать веса до тех пор, пока **ошибка между всеми вычисленными значениями и эталонными значениями не станет минимальной** при одном определенном наборе коэффициентов.

Для этого можем ввести **функцию потерь** и постараться ее минимизировать, то есть минимизировать разницу между предсказаниями и реальными данными. Тогда наша модель будет угадывать точнее.

## Метод наименьших квадратов и функция потерь

Для минимизации функции потерь как один из вариантов используется **метод наименьших квадратов** (Ordinary Least Squares, OLS). Он позволяет найти оптимальные значения коэффициентов линейной регрессии, которые минимизируют квадратичную ошибку.

Задача метода наименьших квадратов заключается в нахождении таких коэффициентов $w_0, w_1, ..., w_d$, которые минимизируют **сумму квадратов ошибок** ($RSS$, risidual sum of squares, или $SSE$ - sum of squared errors) — это одна из функций потерь для регрессии:

$$ RSS = \sum_{i=1}^N (y_i - \hat{y_i})^2 = \sum_{i=1}^N (y_i - (w_0 + w_1x_{i1} + w_2x_{i2} + ... + w_dx_{id}))^2$$

Эту формуму можно трактовать так: SSE - это сумма квадратов разницы между эталонными значениями, обозначенных как $y_i$, и соответствующих предсказанных значений $\hat{y_i}$. То есть мы выбираем строку из датасета с его значениями признаков и целевым значением, значения признаков подставляем в уравнение регрессии. Перемножение значений признаков на коэффициенты регрессии и будет предсказанием. Полученное предсказание вычитаем из целевого и возводим в квадрат, это и будет квадрат ошибки. Аналогичное повторяем для всех строк в датасете, и суммируем в итоговую ошибку. Цель - минимизировать эту итоговую ошибку.

Может возникнуть вопрос: а зачем возводить в квадрат? На это есть несколько ответов:

1. Если просто суммировать разности между фактическими и предсказанными значениями, то **положительные ошибки** (когда предсказание меньше фактического значения) **будут компенсироваться отрицательными** (когда предсказание больше фактического значения). Это привело бы к тому, что сумма ошибок могла бы быть близка к нулю, даже если предсказания модели будут очень неточными.

2. Квадрат ошибки делает **большие ошибки значительно более значимыми** по сравнению с маленькими. Это важно для обучения модели, так как крупные ошибки сильнее "наказываются" в функции потерь. Это помогает модели лучше подгоняться под те данные, где расхождения наиболее велики.

3. **Модуль — вычислительно дорогая операция**. Возведение в квадрат даёт гладкую функцию, которую легко оптимизировать с помощью методов линейной алгебры и градиентного спуска. В последующих занятиях будет более подробный разбор градиентного спуска, в этом занятии затронем его поверхностно. На данный момент всё, что нам надо о нем знать - это то, что он помогает находить веса моделей, отыскивая минимум функции потерь, и для его использования используется дифференцирование. Функции, содержащие квадраты, хорошо себя ведут при нахождении минимумов, так как их производные легко вычисляются.

Геометрически метод наименьших квадратов можно интерпретировать как проекцию вектора $y$ (вектора истинных значений) на пространство, натянутое на столбцы матрицы $X$ (признаковое пространство). Это означает, что мы ищем такие коэффициенты $w$, которые минимизируют расстояние между фактическими значениями и гиперплоскостью, описанной моделью линейной регрессии.

Геометрическая интерпретация метода наименьших квадратов (OLS) для одной переменной — это более простая версия многомерного случая и может быть легко визуализирована на плоскости.

Предположим, у нас есть зависимость между одной переменной $x$ (признак) и целевой переменной $y$ (результат). Задача состоит в том, чтобы найти прямую линию (линейную функцию), которая наилучшим образом описывает зависимость $y$ от $x$.

**Каждое наблюдение данных** можно представить как **точку на двумерной плоскости**, где ось $x$ — это значение признака, а ось $y$ — это значение целевой переменной. Например, если у вас есть набор данных, где $x$ — это площадь квартиры, а $y$ — цена квартиры, то каждая точка на графике представляет цену квартиры при определённой площади. В геометрическом смысле мы **ищем линию, которая минимизирует расстояние между фактическими точками данных и предсказанными значениями**.

Для каждой точки данных существует вертикальное отклонение (ошибка) от фактического значения до предсказанного значения на линии. Это отклонение называется остатком (или ошибкой). Геометрически, это расстояние по вертикали от точки фактического значения до линии регрессии. **Метод наименьших квадратов (OLS) заключается в минимизации суммы квадратов этих вертикальных отклонений**.

**Многомерный случай**

Если у нас есть несколько предикторов (независимых переменных), то данные можно представить как точки в многомерном пространстве, где каждая точка имеет координаты, соответствующие значениям предикторов для одного наблюдения.

> Например, если у нас два предиктора ($x_1$ и $x_2$), мы можем визуализировать данные как точки на трехмерной плоскости ($x_1$, $x_2$ и $y$, где $y$ — это зависимая переменная).

Модель линейной регрессии в этом случае ищет **гиперплоскость**, которая лучше всего приближает эти точки.  Гиперплоскость — это многомерное пространство, которое на одну единицу меньше, чем пространство, в котором оно содержится. Если $x_1$ и $x_2$ — это независимые переменные, тогда предсказанные значения $y$ лежат на плоскости в трехмерном пространстве. Для трёх предикторов модель будет пытаться найти гиперплоскость в четырёхмерном пространстве, и так далее.

Каждая точка данных в пространстве соответствует одному наблюдению. Фактические значения $y$ для каждой точки могут не лежать на гиперплоскости, поскольку модель не идеальна и данные могут быть шумными. В этом случае возникает ошибка предсказания — это расстояние между фактической точкой и её проекцией на гиперплоскость.

## Аналитическое решение OLS

Существует готовое аналитическое решение OLS, записанное в матричной форме:

$$ X^TXw = X^Ty$$

Решение этих уравнений даёт:

$$ w = (X^TX)^{-1}X^Ty$$

$w$  – искомые веса, или коэффициенты, линейной регрессии

$X$ – матрица фичей

$X^T$ – транспонированная матрица фичей

$y$  – вектор целевых значений

$-1$  – обозначение обратной матрицы

Это решение и есть вектор коэффициентов линейной регрессии, который определяет гиперплоскость, на которую проецируется вектор $y$.

**Рассмотрим подробнее аналитический вывод**

Задача линейной регрессии состоит в нахождении такого вектора $w$, который минимизирует квадратичную ошибку между предсказанными значениями $\hat{y} = Xw$ и наблюдаемыми значениями $y$.

Функция потерь, которую мы минимизируем:

$$ L(w) = || y - Xw||^2 = (y - Xw)^T(y - Xw) $$

Запись $|| y - Xw||$ — это евклидова норма вектора ошибок, которая обозначает расстояние между векторами $y$ (наблюдаемыми значениями) и $Xw$ (предсказанными значениями). Нас интересует минимизация квадрата нормы, то есть $|| y - Xw||^2$. Это то же самое, что и сумма квадратов отклонений (ошибок) между наблюдаемыми и предсказанными значениями:

$$ || y - Xw||^2 = \sum_{i=1}^n{(y_i - (Xw)_i)^2} $$

Смысл этого выражения — это общая ошибка модели, измеряемая как сумма квадратов разностей между фактическими и предсказанными значениями.

Теперь перейдём к выражению $|| y - Xw||^2 = (y - Xw)^T(y - Xw)$. Это алгебраическое представление квадрата нормы через матричное умножение:
* $(y - Xw)^T$ — транспонированный вектор отклонений (строка),
* $(y - Xw)$ —  вектор отклонений (столбец).

Умножение этих двух векторов даёт скаляр — сумму квадратов всех элементов вектора отклонений. Именно это и соответствует евклидовой норме в квадрате.

Для минимизации ошибки необходимо найти частные производные функции потерь по каждому параметру $w$ и приравнять их к нулю. Для этого:

0. Прежде чем вычислять градиент, раскроем скобки с использованием свойств транспонирования и умножения матриц:

$$ L(w) = (y - Xw)^T(y - Xw) = y^Ty - 2w^TX^Ty + w^TX^TXw$$

1. Запишем выражение для градиента функции потерь по $w$:

$$ \frac{\partial L(w)}{\partial w} = -2X^Ty + 2x^TXw = -2X^T (y - Xw)$$

2. Приравниваем градиент к нулю для нахождения минимума:

$$-2X^T (y - Xw) = 0$$

3. Преобразуем уравнение:

$$ X^Ty = X^TXw$$

4. Решаем относительно $w$:

$$ w = (X^TX)^{-1}X^Ty$$

Это решение существует, если матрица $X^TX$ обратима. Если матрица необратима или плохо обусловлена (когда её детерминант близок к нулю), то возникают проблемы с вычислением обратной матрицы.

## Линейная регрессия на основе градиентного спуска и MSE

Также одной из наиболее распространённых функций потерь является среднеквадратическая ошибка ($MSE$, Mean Squared Error). $MSE$ измеряет среднее значение квадратов отклонений фактических значений от предсказанных моделью:

 $$MSE = \frac{1}{n} \sum_{i=1}^n{(y_i - \hat{y_i})^2}$$

Как можно заметить, между RSS и MSE есть связь — MSE является усреднением RSS:

 $$MSE = \frac{1}{n} \sum_{i=1}^n{(y_i - \hat{y_i})^2} = \frac{1}{n}RSS$$

 Процесс обучения линейной регрессии сводится к минимизации MSE, что означает нахождение таких коэффициентов $w$, которые минимизируют среднеквадратическую ошибку. Частый способ минимизации MSE — использовать итеративный метод, например, **градиентный спуск**. У выбранной функции (в нашем случае MSE) есть свой минимум. И наша цель — с помощью градиентного спуска добраться до этого минимума.

 В случае линейной регрессии, мы используем градиентный спуск, чтобы найти наилучшие коэффициенты линейной модели, которые минимизируют ошибки предсказания.  Градиентный спуск помогает медленно корректировать коэффициенты модели так, чтобы MSE уменьшался. Это делается шаг за шагом. Представьте, что MSE — это рельефная горка, а наша цель — добраться до её подножья, где ошибка минимальна. **Градиентный спуск помогает понять, в каком направлении "спускаться"**, то есть указывает, как нужно изменить коэффициенты, чтобы ошибка уменьшалась. Этот процесс продолжается: мы изменяем коэффициенты модели, пересчитываем MSE и продолжаем корректировать, пока не найдём те, которые минимизируют MSE.

## Интерпретация модели

Интерпретация модели линейной регрессии относительно проста. Основой модели выступают ее коэффициенты $w$: они отражают вклад каждой независимой переменной в предсказание целевой переменной.

* $w_0$  — это свободный член или интерсепт, который показывает значение целевой переменной $y$, когда все независимые переменные равны нулю. Этот коэффициент помогает сместить линию регрессии вдоль оси $y$, чтобы лучше соответствовать данным.

* $w_1, w_2, ..., w_d$ — это коэффициенты наклона, которые показывают, на сколько изменится целевая переменная $y$, если соответствующая независимая переменная $x_1, x_2, .., x_d$ изменится на единицу при условии, что все остальные переменные остаются неизменными.

> Пример: Если у нас есть модель для прогноза цены дома и коэффициент при переменной "площадь" равен 100, это означает, что при увеличении площади дома на 1 квадратный метр, цена дома увеличивается на 100 условных единиц (при прочих равных условиях).

## Недостатки и особенности линейных моделей

Модели линейной регрессии — это одни из самых простых и часто используемых методов в машинном обучении, особенно для задач прогнозирования и анализа взаимосвязей между переменными. Однако, несмотря на свою популярность, у них есть ограничения и особенности, которые важно учитывать.

1. Предполагается, что между признаками и целевой переменной существует линейная зависимость. Если это не так, то модель не будет эффективно работать. Верно и обратное: линейная регрессия плохо работает на данных с нелинейными зависимостями.

2. Если независимые переменные сильно коррелированы между собой (мультиколлинеарность), это может привести к нестабильным коэффициентам регрессии и затруднить интерпретацию модели.

3. Линейная регрессия сильно подвержена влиянию выбросов. Одно или несколько аномальных значений могут значительно сместить линию регрессии.

Однако из-за своей простоты модель имеет и плюсы:

1. Линейная регрессия требует небольшого объема вычислений, что делает ее подходящей для быстрого анализа данных и начальных этапов исследования.

2. Одно из главных преимуществ линейной регрессии заключается в том, что каждый коэффициент имеет четкое интерпретационное значение, показывая, на сколько изменится целевая переменная при изменении данной фичи на единицу.

# Регуляризация

Когда модель слишком хорошо подстраивается под обучающие данные, она может "переучиться" (переобучиться) и начать делать ошибки на новых данных. Это похоже на ситуацию, когда вы учитесь бросать мяч в одну и ту же корзину с одного и того же места, но как только тебе нужно бросить с другого места или в другую корзину, следует промах.

В линейной регрессии мы строим уравнение, чтобы предсказать результат на основе некоторых признаков. Например, предсказать цену дома на основе его площади, количества комнат и т.д. Иногда модель может присваивать слишком большие коэффициенты этим признакам, чтобы точно подстроиться под обучающие данные. Это и есть переобучение.

Регуляризация добавляет "штраф" к этим коэффициентам (или весам), чтобы модель не могла сделать их слишком большими. Это похоже на то, как если бы тренер давал небольшое утяжеление на руку, чтобы броски были более точными и стабильными.

Если говорить более формально, регуляризация — это методика в машинном обучении, направленная на уменьшение переобучения модели за счёт добавления дополнительного ограничения или штрафа к функции потерь. Основная идея состоит в том, чтобы сделать модель более простой и менее склонной к подгонке под случайные шумы в данных.

## L1 (LASSO)

L1-регуляризация также известна как Lasso (Least Absolute Shrinkage and Selection Operator) регуляризация. Она основана на добавлении штрафа, равного сумме абсолютных значений коэффициентов модели.



$$ \sum_{i=1}^n(y_i - \hat{y_i})^2 + \lambda ||w||_1  → min$$

где:
$\sum_{i=1}^n(y_i - \hat{y_i})^2$  — выбранная функция потерь, которую минимизируем, $w = (w_1, ..., w_d)$ — вектор весов.

Но никто же не мешает вам расписать предсказание $ \hat{y_i}$ как: $$ \hat{y_i} = w_0 + \sum_{j=1}^d {w_j x_{ij}}$$ откуда и появляются веса. То есть расписали предсказание как сумму весов, умноженных на значения признаков.

Теперь распишем сам регуляризирующий член как сумму абсолютных значений коэффициентов (согласно формулировке):

$$λ||w||_1 = \lambda\sum_{j = 1}^{d} |w_j|$$  - L1-регуляризация, штраф

$λ$ - гиперпараметр, который позволяет контролировать, насколько штрафуем веса. Чем больше лямбда, тем сильнее модель штрафуется за величину коэффициентов и их количество. Если занулить лямбду, мы получим самую обычную функцию потерь методом наименьших квадратов, соответственно — самую обычную регрессию.

> Оба способа регуляризации приводят к уменьшению параметров модели, заставляя модель пытаться уменьшить эту сумму.
Однако этот штраф помогает предотвратить переобучение и может даже полностью исключить несущественные признаки, делая веса равными нулю, и **регуляризация Лассо имеет тенденцию занулять коэффициенты сильнее**, чем L2-регуляризация.

## L2 (Ridge)

Аналогичное можно расписать и для ридж-регрессии. L2-регуляризация основана на добавлении штрафа, равного сумме квадратов значений коэффициентов модели.


$$\sum_{i=1}^n(y_i - \hat{y_i})^2 + \lambda ||w||_2^2 → min $$

где: $\sum_{i=1}^n(y_i - \hat{y_i})^2$  - ваша функция потерь, которую минимизируем,
$x = (x_1,... , x_d)$ — вектор признаков, $w = (w_1, ..., w_d)$ — вектор весов.

$$λ||w||_2^2 = \lambda\sum_{j = 1}^{d} w_j^2$$  - L2-регуляризация, штраф

$λ$ - гиперпараметр, который позволяет контролировать, насколько штрафуем веса.

Однако регрессия Ридж имеет тенденцию занижать значения коэффициентов, но не полностью сводить их к нулю.


## Геометрическая интерпретация

Когда мы решаем задачу оптимизации в машинном обучении, основная цель — минимизировать функцию потерь. Чем лучше мы это сделаем, тем ближе мы к оптимальному решению, то есть к минимальному значению функции. Теперь давайте посмотрим, как регуляризация влияет на этот процесс с геометрической точки зрения.

В нашем случае функция потерь — это RSS (сумма квадратов отклонений), которую нужно минимизировать. Чтобы проще понять ситуацию, представим, что у нас есть всего один признак в данных. $x_0$ — это фиктивный признак, который создали искусственно.

На графике показано решение $\hat{w}$, которое является эталонным, но получено без учета регуляризации. Это так называемый "идеальный" минимум для конкретной функции потерь.

Функция двух переменных (двух коэффициентов) образует поверхность в трехмерном пространстве. Значение функции в каждой точке $(w_1, w_2)$ определяет высоту этой поверхности — значение ошибки. Если мы посмотрим сверху на эту поверхность, то увидим контурные линии, которые соответствуют фиксированным значениям функции. Красные эллипсы на графике иллюстрируют вид функции потерь сверху — как если бы мы рассматривали срезы уровней. Это похоже на топографические карты, где линии показывают высоту, только здесь они указывают направление к минимуму.

Теперь о регуляризации. Регуляризация вводит штрафы или ограничения для нашей модели. Геометрически это означает, что все коэффициенты, которые выходят за пределы заданной формы (круга или квадрата), наказываются. Эти формы определяются типом регуляризации — L1 (Lasso) или L2 (Ridge) — и задают ограничения на сумму коэффициентов. Если эта сумма превышает установленный порог, срабатывает коэффициент $λ$, который резко увеличивает значение функции потерь, делая решение неоптимальным.

Чтобы решение считалось оптимальным с учетом регуляризации, оно должно находиться внутри области, задаваемой регуляризацией. В противном случае, как только решение выходит за пределы этой области, коэффициент $λ$ значительно увеличивает функцию потерь. Это заставляет алгоритм корректировать параметры, чтобы найти компромиссное решение, минимизирующее потери и остающееся в пределах регуляризационной области.

Если будем искать оптимальные коэффициенты (веса) модели, то оптимальные решения находятся внутри круга (квадрата) и на его границах. ## ElasticNet

ElasticNet - это комбинация из Ридж- и Лассо-регрессии.

В более привычной форме его можно записать так:

$$(\lambda_1 \sum|w_j| + \lambda_2 \sum w_j^2)$$

У этой регуляризации есть настроиваемый гиперпараметр — l1_ratio, указывающий на долю L1 и L2 регуляризации. l1_ratio = 0, то используется только L2 регуляризация, при l1_ratio = 1 - только L1. При 0 < l1_ratio < 1 - комбинация L1 и L2.

# Полиномиальная регрессия

**Полиномиальная регрессия** — расширение линейной регрессии, позволяющее моделировать нелинейные связи за счет добавления полиномиальных признаков (квадрат, куб и т.д.):

$y_{pred}(x)=w_0+\sum_{j = 1}^{d} w_j x_j^k, k = \overline{1, n}$

$n$ - степень полинома

$w$ - веса/коэффициенты

Простая линейная регрессия представляет собой не что иное, как полиномиальную регрессию первого порядка. В отличие от линейной регрессии, где предполагается, что зависимость между признаками и целевой переменной является линейной, полиномиальная регрессия может описывать кривые, благодаря использованию полиномиальных степеней признаков.

Может возникнуть вопрос, а как тогда искать коэффициенты при новых полиномиальных признаках? На самом деле, решение не будет сильно отличаться. Новые полиномиальные признаки созданы искусственно, однако об этой "искусственности" знают только создатели. Сама модель регрессии о полиномиальности ничего не знает: так, для неё, например, созданный признак $x_1^2$ равнозначен новому вводимому $x_{d+1}$, а о квадратичности преобразования регрессия ничего не знает. Если посмотрим на формулу, то при весах нет степени полинома - степенью наделяются исключительно признаки $x$.

**С увеличением степени полинома и количества признаков размерность задачи быстро возрастает**, что может приводить к "проклятию размерности". Это особенно актуально при использовании моделей с многими признаками, так как количество полиномиальных комбинаций быстро увеличивается.

Одна из проблем полиномиальной регрессии — склонность к переобучению при использовании полиномов высокой степени. Модель может идеально подогнаться под обучающую выборку, но плохо обобщаться на новых данных. Это происходит из-за слишком большой гибкости модели, которая начинает "запоминать" шум данных вместо того, чтобы выявлять общие закономерности. Чтобы избежать этого, важно контролировать степень полинома и использовать регуляризацию.

Интерпретация полиномиальной регрессии сложнее по сравнению с линейной регрессией. Каждое изменение степени полинома меняет форму зависимости, и это делает анализ влияния отдельных признаков менее очевидным.

# Метрики

Метрики VS функция потерь:

* функция потерь оптимизируется/минимизируется в процессе обучения

* метрики служат для оценки качества/производительности модели

Метрика качества — это оценка качества нашей модели. Причем для конкретной задачи может возникать целая иерархия метрик. Это внешний критерий качества, зависящий от предсказанных значений, но не от параметров модели.

Функция потерь — критерий «обучаемости» нашей модели. Возникает, когда мы переходим от построения модели к задаче оптимизации, то есть поиску оптимальных параметров модели.



 $$ MSE = \frac{1}{n} \sum_{i=1}^n(y_i - \hat{y_i})^2 $$

Её мы уже затрагивали - среднеквадратичная ошибка (Mean Squared Error), иначе - средняя сумма квадратов отклонений предсказанных значений от эталонных значений из датасета. Чем ближе она к нулю, тем лучше модель адаптировалась под данные.

MSE отличается от других метрик регрессии своей чувствительностью к большим ошибкам благодаря квадратичному штрафу, который он накладывает. В отличие от метрик, таких как MAE (Mean Absolute Error), которая измеряет среднюю абсолютную ошибку, MSE возводит каждую ошибку в квадрат, что делает его более чувствительным к крупным отклонениям.

Эта квадратичная природа MSE приводит к тому, что большие ошибки оказывают значительно большее влияние на итоговую оценку, чем маленькие ошибки. Это свойство полезно, когда важно минимизировать крупные ошибки, так как модель, оптимизированная по MSE, будет стремиться избегать значительных отклонений.

Однако эта же особенность делает MSE менее подходящим в ситуациях, где данные содержат выбросы. Выбросы могут существенно увеличить значение MSE, что усложняет объективное сравнение моделей и может привести к тому, что модель будет чрезмерно настроена на минимизацию влияния выбросов, игнорируя при этом основное распределение данных.

 $$ RMSE = \sqrt{MSE} $$

RMSE — корень из среднеквадратичной ошибки. Основное преимущество RMSE заключается в том, что она измеряет ошибку в тех же единицах, что и зависимая переменная, что делает её легко интерпретируемой. Более того, RMSE дает больший вес большим ошибкам из-за операции возведения в квадрат и последующего извлечения корня. Однако как и любая метрика, RMSE не лишена недостатков. Например, она также может быть чувствительна к выбросам, поскольку большие отклонения могут значительно увеличивать значение RMSE и искажать общую оценку качества модели.


 $$ MAE = \frac{1}{n} \sum_{i=1}^n|y_i - \hat{y_i}| $$

— усредненная сумма разницы между эталонным и предсказанным значением, взятой по модулю.

MAE (Mean Absolute Error - средняя абсолютная ошибка) отличается тем, что использует абсолютные значения ошибок, а не их квадраты. Это делает MAE менее чувствительным к выбросам и крупным отклонениям, поскольку каждая ошибка вносит линейный вклад в общую оценку, независимо от её величины.

Основная особенность MAE заключается в том, что она равномерно штрафует все ошибки, без дополнительного веса для больших ошибок, как это делает MSE. Это делает MAE более устойчивой к выбросам, поскольку крупные ошибки не оказывают чрезмерного влияния на итоговый результат. Благодаря этому свойству MAE лучше подходит для ситуаций, где данные содержат выбросы, и важно оценить среднюю ошибку без значительного искажения из-за редких, но больших ошибок.

$$ MAPE = \frac{1}{n} \sum_{i=1}^n\left|{\frac{y_i - \hat{y_i}}{y_i}}\right| $$

Используется, чтобы учитывать масштаб целевой переменной. То есть мы минимизируем процент,на который наша модель ошибается.

MAPE — метрика с очень простой интерпретацией, измеряемая в долях или процентах. Если у вас получилось, например, что MAPE = n%, то это говорит о том, что ошибка составила n% от реальных значений. Однако такая метрика нестабильна: любое сильное отклонение может значительно повлиять на размер ошибки.

$$ R^2 = 1 - \frac{\sum_{i=1}^n(y_i - \hat{y_i})^2}{\sum_{i=1}^n(y_i - \overline{y})^2} $$

$R^2 - $ коэффициент детерминации, показывает, какую часть изменчивости наблюдаемой переменной можно объяснить с помощью построенной модели. В отличие от других метрик, эту мы стремимся максимизировать при обучении модели, т. к. она максимальна при MSE = 0, например. Значение R2 меньше 0 означает, что рассматриваемая модель хуже константой. У идеального решающего правила $ R^2 = 1$.

Легко интерпретировать — определяет долю изменений, обусловленных влиянием признаков на целевую переменную. Однако не уменьшается при добавлении новых признаков, независимо от их связи с целевой переменной.
