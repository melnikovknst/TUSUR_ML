## –í–≤–µ–¥–µ–Ω–∏–µ

–í –º–∏—Ä–µ –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è –∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –∑–∞–Ω–∏–º–∞–µ—Ç –æ—Å–æ–±–æ–µ –º–µ—Å—Ç–æ. –≠—Ç–æ –ø—Ä–æ—Å—Ç–æ–π, –Ω–æ –º–æ—â–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è –≤ —Å–∞–º—ã—Ö —Ä–∞–∑–Ω—ã—Ö –æ–±–ª–∞—Å—Ç—è—Ö, –æ—Ç –º–µ–¥–∏—Ü–∏–Ω—ã –¥–æ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–∞. –í –æ—Ç–ª–∏—á–∏–µ –æ—Ç –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç —á–∏—Å–ª–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º—É –∫–ª–∞—Å—Å—É. –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç —Ä–µ—à–∞—Ç—å —Ç–∞–∫–∏–µ –∑–∞–¥–∞—á–∏, –∫–∞–∫ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∑–∞–±–æ–ª–µ–≤–∞–Ω–∏—è —É –ø–∞—Ü–∏–µ–Ω—Ç–∞ –∏–ª–∏ –æ—Ü–µ–Ω–∫–∞ —à–∞–Ω—Å–æ–≤ —Ç–æ–≥–æ, —á—Ç–æ –∫–ª–∏–µ–Ω—Ç —Å–¥–µ–ª–∞–µ—Ç –ø–æ–∫—É–ø–∫—É.



## –û—Å–Ω–æ–≤–Ω—ã–µ –ø—Ä–∏–Ω—Ü–∏–ø—ã —Ä–∞–±–æ—Ç—ã –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏

–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –æ—Å–Ω–æ–≤–∞–Ω–∞ –Ω–∞ –∫–æ–Ω—Ü–µ–ø—Ü–∏–∏ –ª–∏–Ω–µ–π–Ω–æ–π –º–æ–¥–µ–ª–∏, –≥–¥–µ –∫–∞–∂–¥–∞—è –≤—Ö–æ–¥–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –∏–º–µ–µ—Ç —Å–≤–æ–π –≤–µ—Å –∏–ª–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç, –æ—Ç—Ä–∞–∂–∞—é—â–∏–π –µ—ë –≤–ª–∏—è–Ω–∏–µ –Ω–∞ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é. –ù–æ, –≤ –æ—Ç–ª–∏—á–∏–µ –æ—Ç –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç 0 –¥–æ 1. –≠—Ç–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∏–ª–∏ —Å–∏–≥–º–æ–∏–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π –∏ –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

–ó–¥–µ—Å—å $ z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n $ ‚Äî —ç—Ç–æ –ª–∏–Ω–µ–π–Ω–∞—è –∫–æ–º–±–∏–Ω–∞—Ü–∏—è –≤—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –†–µ–∑—É–ª—å—Ç–∞—Ç –ª–∏–Ω–µ–π–Ω–æ–π –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ $z$ –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç—Å—è —á–µ—Ä–µ–∑ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫—É—é (—Å–∏–≥–º–æ–∏–¥–∞–ª—å–Ω—É—é) —Ñ—É–Ω–∫—Ü–∏—é, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –º–æ–¥–µ–ª—å –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ –æ–¥–Ω–æ–º—É –∏–∑ –¥–≤—É—Ö –∫–ª–∞—Å—Å–æ–≤. –ï—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –≤—ã—à–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ –ø–æ—Ä–æ–≥–∞, –Ω–∞–ø—Ä–∏–º–µ—Ä 0,5, —Ç–æ –æ–±—ä–µ–∫—Ç –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –∫ –∫–ª–∞—Å—Å—É 1, –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ ‚Äî –∫ –∫–ª–∞—Å—Å—É 0.

–ü—Ä–æ—Ü–µ—Å—Å –æ–±—É—á–µ–Ω–∏—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –≤–∫–ª—é—á–∞–µ—Ç –æ—Ü–µ–Ω–∫—É –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏ (–≤–µ—Å–æ–≤ $ \beta $), –∫–æ—Ç–æ—Ä—ã–µ –º–∞–∫—Å–∏–º–∏–∑–∏—Ä—É—é—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å (–ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏–µ) –Ω–∞–±–ª—é–¥–∞–µ–º—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –≠—Ç–æ—Ç –ø–æ–¥—Ö–æ–¥ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –º–µ—Ç–æ–¥–æ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è –∏ —è–≤–ª—è–µ—Ç—Å—è –æ–¥–Ω–∏–º –∏–∑ —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã—Ö –∫–æ–Ω—Ü–µ–ø—Ç–æ–≤ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏.



## –ú–∞–∫—Å–∏–º–∏–∑–∞—Ü–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è

–ß—Ç–æ–±—ã –æ–±—É—á–∏—Ç—å –º–æ–¥–µ–ª—å, –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ —Ç–∞–∫–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã  $ùõΩ$, –∫–æ—Ç–æ—Ä—ã–µ –º–∞–∫—Å–∏–º–∏–∑–∏—Ä—É—é—Ç —Ñ—É–Ω–∫—Ü–∏—é –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è. –§–æ—Ä–º—É–ª–∞ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è, –Ω–µ—Å–º–æ—Ç—Ä—è –Ω–∞ –µ—ë –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –≤–∏–¥, –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ –¥–æ–≤–æ–ª—å–Ω–æ –∏–Ω—Ç—É–∏—Ç–∏–≤–Ω–∞. –û–Ω–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ö–æ—Ä–æ—à–æ –Ω–∞—à–∞ –º–æ–¥–µ–ª—å —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –æ–±—ä—è—Å–Ω—è–µ—Ç –Ω–∞–±–ª—é–¥–∞–µ–º—ã–µ –¥–∞–Ω–Ω—ã–µ.

–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ, —á—Ç–æ —É –Ω–∞—Å –µ—Å—Ç—å –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å (–Ω–∞–ø—Ä–∏–º–µ—Ä, –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è), –∫–æ—Ç–æ—Ä–∞—è –ø—ã—Ç–∞–µ—Ç—Å—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç—å —á—Ç–æ-—Ç–æ, –Ω–∞–ø—Ä–∏–º–µ—Ä, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —á–µ–ª–æ–≤–µ–∫ –∫—É–ø–∏—Ç —Ç–æ–≤–∞—Ä (1 ‚Äî –∫—É–ø–∏—Ç, 0 ‚Äî –Ω–µ –∫—É–ø–∏—Ç). –ú–æ–¥–µ–ª—å –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è.

–§—É–Ω–∫—Ü–∏—è –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è ‚Äî —ç—Ç–æ —Å–ø–æ—Å–æ–± –∏–∑–º–µ—Ä–∏—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ "–ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–Ω–æ", —á—Ç–æ –Ω–∞—à–∏ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è (—Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–∫—É–ø–∫–∏ –∏–ª–∏ –∏—Ö –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ) –º–æ–≥–ª–∏ –±—ã –ø—Ä–æ–∏–∑–æ–π—Ç–∏ –ø—Ä–∏ —ç—Ç–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–∞—Ö –º–æ–¥–µ–ª–∏.


$$ L(\beta) = \prod_{i=1}^n P(y_i|X_i,\beta) = \prod_{i=1}^n P_{i}^{y_i} (1-P_i)^{1-y_i}$$

* $P_i$  ‚Äî —ç—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –¥–ª—è –Ω–∞–±–ª—é–¥–µ–Ω–∏—è $i$ –º–æ–¥–µ–ª—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–ª–∞ –∫–ª–∞—Å—Å 1 (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–∫—É–ø–∫–∏)
* $y_i$  ‚Äî –∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç–∫–∏ –¥–ª—è —ç—Ç–æ–≥–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è ($y_i=1$, –µ—Å–ª–∏ –ø–æ–∫—É–ø–∫–∞ –ø—Ä–æ–∏–∑–æ—à–ª–∞, –∏ $y_i=0$, –µ—Å–ª–∏ –ø–æ–∫—É–ø–∫–∞ –Ω–µ –ø—Ä–æ–∏–∑–æ—à–ª–∞)
* $1‚àíP_i$  ‚Äî –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –º–æ–¥–µ–ª—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–ª–∞ –∫–ª–∞—Å—Å 0 (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ø–æ–∫—É–ø–∫–∞ –Ω–µ —Å–æ—Å—Ç–æ—è–ª–∞—Å—å).

$P_i^{y_i}$ ‚Äî –≤—ã—Ä–∞–∂–µ–Ω–∏–µ —É—á–∏—Ç—ã–≤–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è, –µ—Å–ª–∏ –¥–ª—è –Ω–∞–±–ª—é–¥–µ–Ω–∏—è $i$ $y_i=1$. $(1-P_i^{1-y_i})$ ‚Äî –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ–µ –¥–ª—è $y_i=0$.

* –ï—Å–ª–∏ –º–æ–¥–µ–ª—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–ª–∞, —á—Ç–æ —Å–æ–±—ã—Ç–∏–µ –ø—Ä–æ–∏–∑–æ—à–ª–æ ($y_i=1$), –æ–Ω–∞ "–Ω–∞–≥—Ä–∞–∂–¥–∞–µ—Ç—Å—è" –∑–∞ —ç—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é $P_i$, —Ç–∞–∫ –∫–∞–∫ –≤–æ –≤—Ç–æ—Ä–æ–π —á–∞—Å—Ç–∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è –∏–∑-–∑–∞ –º–∏–Ω—É—Å–∞ –≤ —Å—Ç–µ–ø–µ–Ω–∏ –¥–∞—Å—Ç –Ω—É–ª–µ–≤—É—é —Å—Ç–µ–ø–µ–Ω—å, –∞ –ª—é–±–æ–µ —á–∏—Å–ª–æ –≤ —Å—Ç–µ–ø–µ–Ω–∏ –Ω–æ–ª—å –¥–∞–µ—Ç –µ–¥–∏–Ω–∏—Ü—É.
* –ï—Å–ª–∏ –º–æ–¥–µ–ª—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–ª–∞, —á—Ç–æ —Å–æ–±—ã—Ç–∏–µ –Ω–µ –ø—Ä–æ–∏–∑–æ—à–ª–æ ($y_i=0$), –æ–Ω–∞ "–Ω–∞–≥—Ä–∞–∂–¥–∞–µ—Ç—Å—è" –∑–∞ —ç—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é $1-P_i$, –∞ –ø–µ—Ä–≤–∞—è —á–∞—Å—Ç—å –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è –ø–æ —Ç–æ–π –∂–µ –ø—Ä–∏—á–∏–Ω–µ –±—É–¥–µ—Ç —Ä–∞–≤–Ω–∞ –µ–¥–∏–Ω–∏—Ü–µ.

–ü–æ—Å–ª–µ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –æ—Ü–µ–Ω–∏—Ç—å –µ—ë –∫–∞—á–µ—Å—Ç–≤–æ. –î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è —Ä–∞–∑–ª–∏—á–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–º–æ–≥–∞—é—Ç –ø–æ–Ω—è—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ç–æ—á–Ω–æ –º–æ–¥–µ–ª—å –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç –∫–ª–∞—Å—Å—ã –∏ –∫–∞–∫ –æ–Ω–∞ —Å–ø—Ä–∞–≤–ª—è–µ—Ç—Å—è —Å —Ä–∞–∑–ª–∏—á–Ω—ã–º–∏ –∞—Å–ø–µ–∫—Ç–∞–º–∏ –∑–∞–¥–∞—á–∏. –í –æ—Ç–ª–∏—á–∏–µ –æ—Ç –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –≥–¥–µ –æ—Å–Ω–æ–≤–Ω–æ–π –º–µ—Ç—Ä–∏–∫–æ–π —è–≤–ª—è–µ—Ç—Å—è —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ (MSE), –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è —Ç—Ä–µ–±—É–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã—Ö –º–µ—Ç—Ä–∏–∫ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º –Ω–∞–∏–±–æ–ª–µ–µ –≤–∞–∂–Ω—ã–µ –∏–∑ –Ω–∏—Ö –∏ —Ä–∞–∑–±–µ—Ä–µ–º –∏—Ö –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –∏ –æ–±–ª–∞—Å—Ç–∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è.


### 1. **–¢–æ—á–Ω–æ—Å—Ç—å (Accuracy)**

–¢–æ—á–Ω–æ—Å—Ç—å ‚Äî —ç—Ç–æ –æ–¥–Ω–∞ –∏–∑ –Ω–∞–∏–±–æ–ª–µ–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω—ë–Ω–Ω—ã—Ö –∏ –∏–Ω—Ç—É–∏—Ç–∏–≤–Ω–æ –ø–æ–Ω—è—Ç–Ω—ã—Ö –º–µ—Ç—Ä–∏–∫. –û–Ω–∞ –∏–∑–º–µ—Ä—è–µ—Ç –¥–æ–ª—é –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫:

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

–≥–¥–µ:
- \( TP \) (True Positives) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å—Ç–∏–Ω–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (–º–æ–¥–µ–ª—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–ª–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å),
- \( TN \) (True Negatives) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å—Ç–∏–Ω–Ω–æ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (–º–æ–¥–µ–ª—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–ª–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å),
- \( FP \) (False Positives) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–æ–∂–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (–º–æ–¥–µ–ª—å –æ—à–∏–±–æ—á–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–ª–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å),
- \( FN \) (False Negatives) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–æ–∂–Ω–æ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (–º–æ–¥–µ–ª—å –æ—à–∏–±–æ—á–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–ª–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å).

–¢–æ—á–Ω–æ—Å—Ç—å —Ö–æ—Ä–æ—à–∞ –≤ —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ –∫–ª–∞—Å—Å—ã —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω—ã, —Ç–æ –µ—Å—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤ –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞ –ø—Ä–∏–º–µ—Ä–Ω–æ –æ–¥–∏–Ω–∞–∫–æ–≤–æ. –û–¥–Ω–∞–∫–æ –ø—Ä–∏ —Ä–∞–±–æ—Ç–µ —Å –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –∫–æ–≥–¥–∞ –æ–¥–∏–Ω –∫–ª–∞—Å—Å –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –≥–æ—Ä–∞–∑–¥–æ —á–∞—â–µ –¥—Ä—É–≥–æ–≥–æ) —ç—Ç–∞ –º–µ—Ç—Ä–∏–∫–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ–±–º–∞–Ω—á–∏–≤–æ–π. –í —Ç–∞–∫–∏—Ö —Å–∏—Ç—É–∞—Ü–∏—è—Ö –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å –≤—ã—Å–æ–∫—É—é —Ç–æ—á–Ω–æ—Å—Ç—å, –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—è –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤ –∫–∞–∫ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–∏–µ –∫ –±–æ–ª–µ–µ —á–∞—Å—Ç–æ–º—É –∫–ª–∞—Å—Å—É, –∏–≥–Ω–æ—Ä–∏—Ä—É—è —Ä–µ–¥–∫–∏–π –∫–ª–∞—Å—Å.



### 2. **–ü–æ–ª–Ω–æ—Ç–∞ (Recall)**

–ü–æ–ª–Ω–æ—Ç–∞ (–∏–ª–∏ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å) –∏–∑–º–µ—Ä—è–µ—Ç, –∫–∞–∫—É—é –¥–æ–ª—é –æ–±—ä–µ–∫—Ç–æ–≤ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ –º–æ–¥–µ–ª—å —Å–º–æ–≥–ª–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç—å. –û–Ω–∞ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫:

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

–ü–æ–ª–Ω–æ—Ç–∞ –≤–∞–∂–Ω–∞ –≤ —Ç–µ—Ö —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–æ–º —è–≤–ª—è–µ—Ç—Å—è –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏–µ –≤—Å–µ—Ö –æ–±—ä–µ–∫—Ç–æ–≤ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞, –¥–∞–∂–µ –µ—Å–ª–∏ –ø—Ä–∏ —ç—Ç–æ–º —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–æ–∂–Ω—ã—Ö —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏–π. –ù–∞–ø—Ä–∏–º–µ—Ä, –≤ –º–µ–¥–∏—Ü–∏–Ω–µ –ø—Ä–∏ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–µ —Ä–µ–¥–∫–∏—Ö –∑–∞–±–æ–ª–µ–≤–∞–Ω–∏–π –≤–∞–∂–Ω–æ –≤—ã—è–≤–∏—Ç—å –∫–∞–∫ –º–æ–∂–Ω–æ –±–æ–ª—å—à–µ –±–æ–ª—å–Ω—ã—Ö, –¥–∞–∂–µ –µ—Å–ª–∏ —ç—Ç–æ –ø—Ä–∏–≤–µ–¥–µ—Ç –∫ –ª–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º –¥–∏–∞–≥–Ω–æ–∑–∞–º.



### 3. **–¢–æ—á–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (Precision)**

–¢–æ—á–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫—É—é –¥–æ–ª—é –∏–∑ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –º–æ–¥–µ–ª—å—é –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç –∏—Å—Ç–∏–Ω–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ. –§–æ—Ä–º—É–ª–∞ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è:

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

–≠—Ç–∞ –º–µ—Ç—Ä–∏–∫–∞ –ø–æ–ª–µ–∑–Ω–∞ –≤ —Å–∏—Ç—É–∞—Ü–∏—è—Ö, –∫–æ–≥–¥–∞ –≤–∞–∂–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π. –ù–∞–ø—Ä–∏–º–µ—Ä, –ø—Ä–∏ —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ —Å–ø–∞–º–∞ –ª—É—á—à–µ, —á—Ç–æ–±—ã –º–æ–¥–µ–ª—å –Ω–µ –±–ª–æ–∫–∏—Ä–æ–≤–∞–ª–∞ –ª–µ–≥–∏—Ç–∏–º–Ω—ã–µ —Å–æ–æ–±—â–µ–Ω–∏—è, –¥–∞–∂–µ –µ—Å–ª–∏ —ç—Ç–æ –ø—Ä–∏–≤–µ–¥–µ—Ç –∫ –ø—Ä–æ–ø—É—Å–∫—É –Ω–µ–∫–æ—Ç–æ—Ä–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–ø–∞–º–∞.



### 4. **F1-–º–µ—Äa**

F1-–º–µ—Ä–∞ –æ–±—ä–µ–¥–∏–Ω—è–µ—Ç –ø–æ–ª–Ω–æ—Ç—É –∏ —Ç–æ—á–Ω–æ—Å—Ç—å –≤ –æ–¥–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ, –ø–æ–∑–≤–æ–ª—è—è —É—á–µ—Å—Ç—å –∏—Ö –±–∞–ª–∞–Ω—Å. –û–Ω–∞ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –∫–∞–∫ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –ø–æ–ª–Ω–æ—Ç—ã:

$$
F1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

F1-–º–µ—Ä–∞ –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ–ª–µ–∑–Ω–∞, –∫–æ–≥–¥–∞ –≤–∞–∂–Ω–æ –Ω–∞–π—Ç–∏ –∫–æ–º–ø—Ä–æ–º–∏—Å—Å –º–µ–∂–¥—É —Ç–æ—á–Ω–æ—Å—Ç—å—é –∏ –ø–æ–ª–Ω–æ—Ç–æ–π, –∏ –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –ï—Å–ª–∏ F1-–º–µ—Ä–∞ –±–ª–∏–∑–∫–∞ –∫ 1, —ç—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –º–æ–¥–µ–ª—å —Ö–æ—Ä–æ—à–æ —Å–ø—Ä–∞–≤–ª—è–µ—Ç—Å—è —Å –∑–∞–¥–∞—á–µ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –æ–±–æ–∏—Ö –∫–ª–∞—Å—Å–æ–≤.



### 5. **–ö—Ä–∏–≤–∞—è ROC –∏ AUC**

ROC-–∫—Ä–∏–≤–∞—è (Receiver Operating Characteristic) –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫ –º–µ–Ω—è–µ—Ç—Å—è –¥–æ–ª—è –∏—Å—Ç–∏–Ω–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (–ø–æ–ª–Ω–æ—Ç–∞) –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –¥–æ–ª–∏ –ª–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö (—Å–ø–µ—Ü–∏—Ñ–∏—á–Ω–æ—Å—Ç—å), –∫–æ–≥–¥–∞ –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –ø–æ—Ä–æ–≥ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏. –ö–∞–∂–¥–∞—è —Ç–æ—á–∫–∞ –Ω–∞ ROC-–∫—Ä–∏–≤–æ–π —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é –ø–æ—Ä–æ–≥–∞.

AUC (Area Under the Curve) ‚Äî —ç—Ç–æ –ø–ª–æ—â–∞–¥—å –ø–æ–¥ ROC-–∫—Ä–∏–≤–æ–π. –ó–Ω–∞—á–µ–Ω–∏–µ AUC –≤–∞—Ä—å–∏—Ä—É–µ—Ç—Å—è –æ—Ç 0 –¥–æ 1 –∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É–µ—Ç –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏: —á–µ–º –±–ª–∏–∂–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫ 1, —Ç–µ–º –ª—É—á—à–µ –º–æ–¥–µ–ª—å —Ä–∞–∑–ª–∏—á–∞–µ—Ç –∫–ª–∞—Å—Å—ã. AUC –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ–ª–µ–∑–µ–Ω –ø—Ä–∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π, —Ç–∞–∫ –∫–∞–∫ –æ–Ω –Ω–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –ø–æ—Ä–æ–≥–æ–≤–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –∏ —É—á–∏—Ç—ã–≤–∞–µ—Ç –≤—Å—é –∫—Ä–∏–≤—É—é —Ü–µ–ª–∏–∫–æ–º.



### 6. **–õ–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∞—è –ø–æ—Ç–µ—Ä—è (Log Loss)**

–õ–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∞—è –ø–æ—Ç–µ—Ä—è –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Å–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –æ—Ç–∫–ª–æ–Ω—è—é—Ç—Å—è –æ—Ç –∏—Å—Ç–∏–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∫–ª–∞—Å—Å–æ–≤. –§–æ—Ä–º—É–ª–∞ Log Loss:

$$
\text{Log Loss} = -\frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(p_i) + (1 - y_i) \log(1 - p_i) \right]
$$

–≥–¥–µ:
- $ N $ ‚Äî –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π,
- $ y_i $ ‚Äî –∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ –¥–ª—è –Ω–∞–±–ª—é–¥–µ–Ω–∏—è $ i $ (0 –∏–ª–∏ 1),
- $ p_i $ ‚Äî –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∫–ª–∞—Å—Å–∞ 1 –¥–ª—è –Ω–∞–±–ª—é–¥–µ–Ω–∏—è $ i $.

Log Loss —à—Ç—Ä–∞—Ñ—É–µ—Ç –º–æ–¥–µ–ª—å –∑–∞ —É–≤–µ—Ä–µ–Ω–Ω—ã–µ, –Ω–æ –æ—à–∏–±–æ—á–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è —Å–∏–ª—å–Ω–µ–µ, —á–µ–º –∑–∞ –Ω–µ–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–µ, —á—Ç–æ –¥–µ–ª–∞–µ—Ç –µ—ë –±–æ–ª–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–π –∫ –∫–∞—á–µ—Å—Ç–≤—É –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π.



### 7. **–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫ (Confusion Matrix)**
