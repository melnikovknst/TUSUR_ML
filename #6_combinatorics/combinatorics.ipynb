#Задача 1

Бросаем кубик 6-гранный.

А - вероятность выпадение 1 или 2.

В = отрицание A.

Найдти величины A и B, распределение дискретной случайной величины X.

```python
# 6 граней с равной вероятностью => 1 = 6p

p = 1/6
A = 2*p
B = 1 - A

# A = P(X=1); B = P(X=0)

E_X_theor = 1*A+0*B

cases = [1,2,3,4,5,6]
size = 10000
X = np.random.choice(cases, size, p = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
X_1 = np.where(np.isin(X, [1, 2]), 1, 0)
E_X = np.mean(X_1)

print(f'Ответ:\n\nA: {A}')
print(f'B: {B}')
print(f"Теоритическое E_X: {E_X_theor}")
print(f"E_X по конечной выборке: {E_X}")
```

# Задача 2

Бросается 2 кубика. Какая вероятность, что выпадет оба раза нечетное число?

```python
# А - первый кубик, Р_А - вероятность выпадения четного
# В - первый кубик, Р_В - вероятность выпадения четного
# P_X - вероятность, оба раза нечетное число

P_A = 1/2
P_B = 1/2

P_X_theor = P_A * P_B

cases = [1,2,3,4,5,6]
size = 10000
X_A = np.random.choice(cases, size, p = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
X_B = np.random.choice(cases, size, p = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

X = (X_A % 2 == 1) & (X_B % 2 == 1)
P_X = np.mean(X)

print(f'Ответ:\n\nТеоритическое P_X: {P_X_theor}')
print(f"P_X по конечной выборке: {P_X}")
```

# Задача 3

 А - выпадение нечетного числа на кубике 1 (неявно, что на втором кубике может выпасть все что угодно)

 B - выпадение 5-ки на кубике 2 (неявно, что на первом кубике может выпасть все что угодно)

 Найти вероятность, что произошло и А и В:




```python
A = 1/2
B = 1/6

P_X_theor = A*B

cases = [1,2,3,4,5,6]
size = 1000
X_A = np.random.choice(cases, size, p = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
X_B = np.random.choice(cases, size, p = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

X = np.where((np.isin(X_A, [1, 3, 5]) & np.isin(X_B, [5])), 1, 0)
P_X = np.mean(X)

print(f'Ответ:\n\nA: {A}')
print(f'B: {B}')
print(f'Теоритическое P_X: {P_X_theor}')
print(f"P_X по конечной выборке: {P_X}")
```
