```python
from tensorflow import keras as k
```

#Задача №1
Разработать структуру распознавания 10 цифр (CNN, размер изображения 28х28x1)

```python
model = k.Sequential()
model.add(k.layers.Conv2D(16, 3, activation = 'relu', input_shape = (28, 28, 1)))
model.add(k.layers.MaxPool2D(pool_size = (2, 2)))
model.add(k.layers.Conv2D(60, 5, activation = 'relu'))
model.add(k.layers.MaxPool2D(pool_size = (2, 2)))
model.add(k.layers.Flatten())
model.add(k.layers.Dense(480, activation = 'relu'))
model.add(k.layers.Dense(10, activation = 'softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy')

model.summary()
```

Данная структра, помтроенная с помощью Sequential API, принимает на вход тензор 28*28*1 и применяет схему свертки *Conv2D* -> *MaxPool2D* -> *повторить*. Это позволяет выделить наиболее общие и информативные черты поданного изображения(за счет слоя *Conv2D*), при этом снижая вычислительные затраты и снижая вероятность переобучения(за счет слоя *MaxPool2D*). После свёрточных слоев, подаем информацию в виде оодномерного массива (*Flatten*) в полносвязные слои. Выходной слой использует функцию активации softmax, тем распределяет вероятности каждого из 10 возможных исходов. При сборке модели была указана функция потерь, характерная для мультиклассовой классификации - *categorical_crossentropy*

#Задача №2
Разработать структуру бинарного классификатора кошек и собак (CNN,  размер изображения 150х150x3)

```python
model = k.Sequential()
model.add(k.layers.Conv2D(32, 3, activation = 'relu', input_shape = (150, 150, 3)))
model.add(k.layers.MaxPool2D(pool_size = (2, 2)))
model.add(k.layers.Conv2D(128, 3, activation = 'relu'))
model.add(k.layers.MaxPool2D(pool_size = (2, 2)))
model.add(k.layers.Conv2D(256, 5, activation = 'relu'))
model.add(k.layers.MaxPool2D(pool_size = (2, 2)))
model.add(k.layers.Flatten())
model.add(k.layers.Dense(1024, activation = 'relu'))
model.add(k.layers.Dense(1, activation = 'sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy')

model.summary()
```

Стратегия аналогична решению предыдущей задачи. Однако, в данном случае был добавлен еще один *Conv2D*+*MaxPool2D* слой и увеличено количество нейронов полносвязного слоя, ввиду большей сложности классификации и большей размерности входных данных. Ключевым отличием является ыункция активация выходного слоя. При задаче бинарной классификации используется логистическая функция - сигмоида и функция потерь, характерная для бинарной классификации - *binary_crossentropy*

#Задача №3
Разработать структуру регрессора, где входные данные представляют вектор (60,1), (MLP)

```python
model = k.Sequential()
model.add(k.layers.Dense(360, activation = 'relu', input_shape = (60,)))
model.add(k.layers.Dense(180, activation = 'relu'))
model.add(k.layers.Dense(1, activation = 'linear'))

model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

Данная нейронная сеть является полносвяной, содержит 3 слоя, чего скорее всего будет достаточно для такой размерности входных данных (количество может варьироваться в зависимости от конкретной задачи). Последний слой содержит 1 нейрон и линейную функцию активации (тк задача регрессии). В качестве функции потерь - MSE.
